```python
# =============================================================================
# CELL 1: Setup, Imports, and Configuration
# =============================================================================
import requests
import time
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
import datetime
from urllib.parse import quote
from tqdm.notebook import tqdm
import re
import json

# Text Processing and ML
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder # For Naive Bayes if target needs encoding
from sentence_transformers import SentenceTransformer, util

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Display
from IPython.display import display, HTML

# --- Configuration ---
XLSX_FILE_PATH = "C:/Users/Galaxy/Downloads/screening_ERAS.xlsx" # !!! UPDATE THIS PATH !!!
NCBI_API_KEY   = "YOUR_NCBI_API_KEY"  # !!! REPLACE with your NCBI API key !!!
CROSSREF_MAILTO= "your_email@example.com" # !!! REPLACE with your email for CrossRef polite pool !!!

# --- NCBI/CrossRef Batch sizes and Delays ---
EFETCH_BATCH_SIZE = 100
ELINK_BATCH_SIZE  = 100
ID_CONVERTER_BATCH_SIZE_CR = 50 # For converting DOIs from CrossRef to PMIDs
CROSSREF_FILTER_BATCH_SIZE = 30 # For fetching refs from CrossRef (URL length limit)
API_DELAY = 0.2 if NCBI_API_KEY and NCBI_API_KEY != "YOUR_NCBI_API_KEY" else 0.4 # Shorter delay if API key is used

# --- Semantic Similarity Model ---
SEMANTIC_MODEL_NAME = 'all-MiniLM-L6-v2' # A good general-purpose model

# --- User-Defined Proposal Query Text ---
# !!! CRITICAL: Define the text that represents YOUR specific research interest/proposal here !!!
# This will be used for semantic similarity calculations.
# Example:
# PROPOSAL_QUERY_TEXT = """
# My research focuses on developing and evaluating an eHealth intervention,
# specifically a mobile application, to support pediatric patients and their families
# through an Enhanced Recovery After Surgery (ERAS) protocol for major airway surgery.
# Key aspects include tracking adherence to perioperative pulmonary rehabilitation,
# managing pain non-pharmacologically, facilitating early mobilization, and measuring
# outcomes like length of stay, functional recovery, and patient satisfaction.
# """
PROPOSAL_QUERY_TEXT = """
Developing and validating an ERAS (Enhanced Recovery After Surgery) protocol
for pediatric airway surgery. This involves creating a mobile application for collecting
anthropometric, clinical, and anesthetic data to inform the ERAS protocol.
The focus is on optimizing perioperative management, including anesthetic techniques,
for different age groups and comorbidity profiles to ensure shorter, more comfortable
post-surgical recovery, reduce adverse events like bronchospasm and desaturation,
and improve overall quality of care and patient safety in procedures such as
tracheoscopies, tracheal dilations, bronchoscopies, and tracheoplasties.
The aim is to generate an evidence-based, app-driven ERAS protocol that can serve
as a model for similar pediatric airway services.
""" # Replace or refine further if needed

# --- Naive Bayes Configuration ---
RELEVANCE_BINS = [0, 3, 7, 10] # (0-3: Low, 4-7: Med, 8-10: High)
RELEVANCE_LABELS = ['Low', 'Medium', 'High']
MIN_SAMPLES_PER_CLASS_NB = 5 # Minimum samples required in at least two classes for NB

# --- TF-IDF Keyword Extraction Configuration ---
NUM_KEYWORDS_TFIDF = 10 # Number of top keywords to extract for keyword-based similarity

# --- Core Set Definition for Structural Metrics ---
# Papers with relevance >= this threshold will form the 'core set' for co-citation/bibliographic coupling
CORE_RELEVANCE_THRESHOLD_FOR_STRUCTURAL = 9

# --- Setup NLTK Stopwords (download if first time) ---
try:
    STOP_WORDS = stopwords.words('english')
except LookupError:
    print("NLTK stopwords not found. Downloading...")
    nltk.download('stopwords')
    STOP_WORDS = stopwords.words('english')

# --- Requests Session (using adapter from your example for robustness) ---
session = requests.Session()
adapter = requests.adapters.HTTPAdapter(pool_connections=50, pool_maxsize=50, max_retries=3)
session.mount("https://", adapter)
session.mount("http://", adapter)

session_cr = requests.Session() # Separate session for CrossRef
adapter_cr = requests.adapters.HTTPAdapter(pool_connections=10, pool_maxsize=10, max_retries=3)
session_cr.mount("https://", adapter_cr)

print("Cell 1: Setup and Configuration Complete.")
print(f"PROPOSAL_QUERY_TEXT set to: '{PROPOSAL_QUERY_TEXT[:100]}...'")

# =============================================================================
# CELL 2: NCBI & CrossRef API Helper Functions
# (Adapted and expanded from your DAG_Checker_V14 and previous script)
# =============================================================================
def _get_ncbi_params(extra=None, tool_name="eras_relevance_v2"):
    """Helper for NCBI params."""
    params = {"tool": tool_name, "email": CROSSREF_MAILTO}
    if NCBI_API_KEY and NCBI_API_KEY != "YOUR_NCBI_API_KEY":
        params["api_key"] = NCBI_API_KEY
    if extra:
        params.update(extra)
    return params

def parse_efetch_article_xml(pubmed_article_node):
    """Parses a single <PubmedArticle> XML node from EFetch."""
    pmid_node = pubmed_article_node.find(".//MedlineCitation/PMID")
    pmid = pmid_node.text if pmid_node is not None else None
    if not pmid: return None

    article_node = pubmed_article_node.find(".//MedlineCitation/Article")
    if article_node is None: return {'PMID_Fetched': pmid} # Basic info if article node missing

    title = article_node.findtext(".//ArticleTitle", default="")
    
    abstract_parts = [at.text for at in article_node.findall(".//Abstract/AbstractText") if at.text]
    abstract = "\n".join(abstract_parts) if abstract_parts else ""

    mesh_headings = []
    for mesh_node in pubmed_article_node.findall(".//MedlineCitation/MeshHeadingList/MeshHeading"):
        desc_name = mesh_node.findtext(".//DescriptorName")
        qual_names = [q.text for q in mesh_node.findall(".//QualifierName")]
        full_mesh = desc_name
        if qual_names:
            full_mesh += "/" + "/".join(qual_names)
        if desc_name: mesh_headings.append(full_mesh)
    
    pub_date = None
    MONTH_MAP = {m: i for i, m in enumerate(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], 1)}
    date_priority = [
        ".//PubmedData/History/PubMedPubDate[@PubStatus='pubmed']", # Most reliable electronic
        ".//PubmedData/History/PubMedPubDate[@PubStatus='entrez']",
        ".//ArticleDate",
        ".//Journal/JournalIssue/PubDate"
    ]
    for path in date_priority:
        date_node = pubmed_article_node.find(path)
        if date_node is not None:
            year, month, day = date_node.findtext("Year"), date_node.findtext("Month"), date_node.findtext("Day")
            try:
                m_val = MONTH_MAP.get(month, int(month) if month and month.isdigit() else 1)
                pub_date = datetime.date(int(year), m_val, int(day))
                break
            except: continue
    
    doi = ""
    # Prioritize ArticleIdList for DOI as it's more standard
    doi_node = pubmed_article_node.find(f".//PubmedData/ArticleIdList/ArticleId[@IdType='doi']")
    if doi_node is not None and doi_node.text:
        doi = doi_node.text
    else: # Fallback to ELocationID (less common for final DOIs)
        for art_id in article_node.findall(".//ELocationID[@EIdType='doi']"):
            if art_id.text:
                doi = art_id.text
                break
    
    journal_title = article_node.findtext(".//Journal/Title", default="")
    
    # Initial (potentially incomplete) list of reference PMIDs from EFetch
    efetch_references_pmids = []
    for ref_node in pubmed_article_node.findall(".//ReferenceList/Reference"):
        ref_pmid_node = ref_node.find(".//ArticleIdList/ArticleId[@IdType='pubmed']")
        if ref_pmid_node is not None and ref_pmid_node.text:
            efetch_references_pmids.append(ref_pmid_node.text)

    return {
        "PMID_Fetched": pmid, "Fetched_Title": title, "Fetched_Abstract": abstract,
        "Fetched_MeSH_Terms": mesh_headings, "Fetched_PubDate": pub_date,
        "Fetched_DOI": doi, "Fetched_Journal": journal_title,
        "Fetched_References_PMIDs_EFetch": efetch_references_pmids # Store EFetch refs separately first
    }

def fetch_pubmed_details_batch(pmids_list):
    """Fetches details for PMIDs using EFetch (POST), including initial refs."""
    all_data = []
    pmids_set = set(pmids_list)
    print(f"Fetching PubMed details for {len(pmids_list)} PMIDs (EFetch)...")
    for i in tqdm(range(0, len(pmids_list), EFETCH_BATCH_SIZE), desc="PubMed EFetch Batches"):
        batch_pmids = pmids_list[i:i + EFETCH_BATCH_SIZE]
        params = _get_ncbi_params({"db": "pubmed", "retmode": "xml", "id": ",".join(batch_pmids)})
        try:
            r = session.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi", data=params, timeout=180)
            r.raise_for_status()
            root = ET.fromstring(r.content)
            for article_xml in root.findall(".//PubmedArticle"):
                parsed_data = parse_efetch_article_xml(article_xml)
                if parsed_data and parsed_data['PMID_Fetched'] in pmids_set:
                    all_data.append(parsed_data)
        except Exception as e: print(f"  - ERROR EFetch batch: {e}")
        time.sleep(API_DELAY)
    df = pd.DataFrame(all_data)
    if not df.empty: df['Fetched_PubDate'] = pd.to_datetime(df['Fetched_PubDate'], errors='coerce').dt.date
    return df

def fetch_pubmed_citations_batch(pmids_list):
    """Fetches PMIDs of citing articles using ELink (POST)."""
    citations_map = {pmid: [] for pmid in pmids_list}
    print(f"Fetching PubMed citations for {len(pmids_list)} PMIDs (ELink)...")
    for i in tqdm(range(0, len(pmids_list), ELINK_BATCH_SIZE), desc="PubMed ELink Batches"):
        batch_pmids = pmids_list[i:i + ELINK_BATCH_SIZE]
        params = _get_ncbi_params({"dbfrom": "pubmed", "db": "pubmed", "linkname": "pubmed_pubmed_citedin", "id": ",".join(batch_pmids)})
        try:
            r = session.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/elink.fcgi", data=params, timeout=120)
            r.raise_for_status()
            root = ET.fromstring(r.content)
            for linkset_node in root.findall(".//LinkSet"):
                source_pmid = linkset_node.findtext("./IdList/Id")
                if source_pmid in citations_map:
                    for link_node in linkset_node.findall(".//LinkSetDb/Link/Id"):
                        if link_node.text: citations_map[source_pmid].append(link_node.text)
        except Exception as e: print(f"  - ERROR ELink batch: {e}")
        time.sleep(API_DELAY)
    return citations_map

# --- CrossRef Augmentation Functions (Adapted from DAG_Checker_V14) ---
def get_crossref_refs_via_filter_batch(doi_list):
    """Queries CrossRef /works using filter=doi:A,doi:B,... to get reference DOIs."""
    if not doi_list: return {}
    print(f"  Fetching reference DOIs from CrossRef for {len(doi_list)} publication DOIs...")
    all_results = {}
    for i in tqdm(range(0, len(doi_list), CROSSREF_FILTER_BATCH_SIZE), desc="CrossRef Ref Batches", leave=False):
        batch_dois = doi_list[i:i + CROSSREF_FILTER_BATCH_SIZE]
        filter_value = ",".join([f"doi:{d.lower()}" for d in batch_dois]) # CR prefers lowercase DOIs
        params = {"filter": filter_value, "rows": len(batch_dois), "select": "DOI,reference"}
        headers = {"User-Agent": f"RelevanceAnalyzer/1.0 (mailto:{CROSSREF_MAILTO})"}
        base_url = "https://api.crossref.org/works"
        try:
            req = requests.Request('GET', base_url, params=params).prepare()
            if len(req.url) > 4000: # Check for overly long URLs
                 print(f"    WARN: URL for CrossRef batch might be too long ({len(req.url)} chars). Consider smaller CROSSREF_FILTER_BATCH_SIZE.")

            r = session_cr.get(base_url, params=params, headers=headers, timeout=120)
            r.raise_for_status()
            data = r.json().get("message", {}).get("items", [])
            for item in data:
                item_doi_cr = item.get("DOI")
                # Match fetched DOI (case-insensitive) to an input DOI from the batch
                matched_input_doi = next((d_in for d_in in batch_dois if d_in.lower() == item_doi_cr.lower()), None)
                if matched_input_doi:
                    refs = item.get("reference", [])
                    ref_dois = [ref.get("DOI") for ref in refs if ref.get("DOI")]
                    all_results[matched_input_doi] = ref_dois
        except Exception as e: print(f"    - ERROR CrossRef batch for DOIs {batch_dois[:2]}...: {e}")
        time.sleep(API_DELAY) # CR has rate limits
    print(f"  CrossRef fetched reference DOIs for {len(all_results)} PMIDs.")
    return all_results

def batch_dois_to_pmids_ncbi(doi_list_to_convert, batch_size=ID_CONVERTER_BATCH_SIZE_CR):
    """Converts a list of DOIs to PMIDs using NCBI ESearch (more robust for batching than ID Converter API)."""
    if not doi_list_to_convert: return {}
    doi_to_pmid_map = {}
    unique_dois = list(set(d.lower() for d in doi_list_to_convert if d)) # Unique, lowercase
    print(f"  Converting {len(unique_dois)} unique reference DOIs to PMIDs via NCBI ESearch...")

    for i in tqdm(range(0, len(unique_dois), batch_size), desc="DOI to PMID (ESearch)", leave=False):
        batch_dois = unique_dois[i:i+batch_size]
        # Constructing a query like: (doi1[DOI] OR doi2[DOI] OR ...)
        query_term = " OR ".join([f"{quote(doi)}[DOI]" for doi in batch_dois])
        params = _get_ncbi_params({
            "db": "pubmed",
            "term": query_term,
            "retmax": str(len(batch_dois)) # Should be enough
        })
        try:
            r = session.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi", data=params, timeout=120)
            r.raise_for_status()
            root = ET.fromstring(r.content)
            # This ESearch directly gives PMIDs. We need to map them back to original DOIs if multiple DOIs map to same PMID or vice-versa.
            # For simplicity here, if ESearch returns PMIDs, we try to associate.
            # A more robust mapping would require EFetch on the found PMIDs to get their DOIs and then match.
            # For now, this is a simpler heuristic.
            pmids_found = [id_node.text for id_node in root.findall(".//IdList/Id")]

            # To map back, we'll fetch details for these PMIDs to get their DOIs
            if pmids_found:
                temp_df_details = fetch_pubmed_details_batch(pmids_found) # Use our existing function
                if not temp_df_details.empty:
                    for _, row in temp_df_details.iterrows():
                        fetched_doi = row['Fetched_DOI']
                        if fetched_doi and fetched_doi.lower() in batch_dois: # Check if this PMID's DOI was in our query batch
                             doi_to_pmid_map[fetched_doi.lower()] = row['PMID_Fetched'] # Map lowercase DOI
                             # Also map original case if different and requested
                             original_case_doi = next((d for d in batch_dois if d.lower() == fetched_doi.lower()), None)
                             if original_case_doi:
                                 doi_to_pmid_map[original_case_doi] = row['PMID_Fetched']


        except Exception as e:
            print(f"    - ERROR ESearch for DOI to PMID batch: {e}")
        time.sleep(API_DELAY)
    print(f"  NCBI ESearch converted {len(doi_to_pmid_map)} DOIs to PMIDs.")
    return doi_to_pmid_map

print("Cell 2: API Helper Functions Defined.")

# =============================================================================
# CELL 3: Load Initial Data & Core PubMed Augmentation
# =============================================================================
print(f"\nLoading data from: {XLSX_FILE_PATH}")
df_main = pd.read_excel(XLSX_FILE_PATH)
df_main['PMID'] = df_main['PMID'].astype(str)
print(f"Loaded {len(df_main)} records from XLSX.")
display(df_main.head(2))

pmids_initial_list = df_main['PMID'].unique().tolist()

# --- Fetch Core PubMed Details (Title, Abstract, MeSH, PubDate, DOI, EFetch Refs) ---
df_pubmed_details = fetch_pubmed_details_batch(pmids_initial_list)
if not df_pubmed_details.empty:
    df_main = pd.merge(df_main, df_pubmed_details, left_on='PMID', right_on='PMID_Fetched', how='left')
    df_main.drop(columns=['PMID_Fetched'], inplace=True, errors='ignore')
else:
    print("Warning: No details fetched from PubMed EFetch.")
    # Add empty columns to prevent errors later if df_pubmed_details is empty
    for col in ["Fetched_Title", "Fetched_Abstract", "Fetched_MeSH_Terms", 
                "Fetched_PubDate", "Fetched_DOI", "Fetched_Journal", 
                "Fetched_References_PMIDs_EFetch"]:
        if col not in df_main.columns: df_main[col] = np.nan if col != "Fetched_MeSH_Terms" else [[] for _ in range(len(df_main))]


# --- Fetch PubMed Citations (Papers that CITE these PMIDs) ---
citations_map = fetch_pubmed_citations_batch(pmids_initial_list)
df_main['Fetched_Cited_By_PMIDs'] = df_main['PMID'].map(citations_map).fillna('').apply(list)


# Ensure list columns are initialized properly
for col in ['Fetched_MeSH_Terms', 'Fetched_References_PMIDs_EFetch', 'Fetched_Cited_By_PMIDs']:
    if col in df_main.columns:
         df_main[col] = df_main[col].apply(lambda x: x if isinstance(x, list) else ([] if pd.isna(x) else [x] if not isinstance(x,list) else [] )) # handle NaNs and ensure list
    else: # if column wasn't created due to empty fetch
        df_main[col] = [[] for _ in range(len(df_main))]


print("\nSample of data after core PubMed augmentation (details & citations):")
display(df_main[['PMID', 'Title', 'Fetched_Title', 'Fetched_DOI', 'Fetched_References_PMIDs_EFetch', 'Fetched_Cited_By_PMIDs']].head())
print("Cell 3: Core PubMed Augmentation Complete.")

# =============================================================================
# CELL 4: Comprehensive Reference Fetching (CrossRef Augmentation)
# =============================================================================
print("\nStarting Comprehensive Reference Fetching via CrossRef...")
# Identify PMIDs that have a DOI, as CrossRef works with DOIs
dois_for_crossref = df_main[df_main['Fetched_DOI'].notna() & (df_main['Fetched_DOI'] != "")]['Fetched_DOI'].unique().tolist()

# 1. Get Reference DOIs from CrossRef for our publication DOIs
if dois_for_crossref:
    pub_doi_to_ref_dois_map = get_crossref_refs_via_filter_batch(dois_for_crossref)

    # 2. Collect all unique reference DOIs that need conversion to PMID
    all_ref_dois_to_convert = set()
    for ref_list in pub_doi_to_ref_dois_map.values():
        all_ref_dois_to_convert.update(ref_list)
    
    # 3. Convert these reference DOIs to PMIDs
    if all_ref_dois_to_convert:
        ref_doi_to_pmid_map = batch_dois_to_pmids_ncbi(list(all_ref_dois_to_convert))
    else:
        ref_doi_to_pmid_map = {}
        print("No reference DOIs found from CrossRef to convert to PMIDs.")

    # 4. Create the final 'Fetched_References_PMIDs_Combined' list
    def combine_references(row):
        efetch_refs = set(row['Fetched_References_PMIDs_EFetch'] if isinstance(row['Fetched_References_PMIDs_EFetch'], list) else [])
        
        crossref_ref_pmids = set()
        pub_doi = row['Fetched_DOI']
        if pd.notna(pub_doi) and pub_doi in pub_doi_to_ref_dois_map:
            for ref_doi in pub_doi_to_ref_dois_map[pub_doi]:
                # Use lowercase for lookup in map, as we stored them lowercase from ESearch results
                ref_pmid = ref_doi_to_pmid_map.get(ref_doi.lower()) 
                if ref_pmid:
                    crossref_ref_pmids.add(ref_pmid)
        
        # Combine and ensure PMIDs are strings and unique
        combined_list = list(set(map(str, efetch_refs.union(crossref_ref_pmids))))
        return [r for r in combined_list if r != row['PMID']] # Exclude self-references

    df_main['Fetched_References_PMIDs_Combined'] = df_main.apply(combine_references, axis=1)
else:
    print("No DOIs found in the dataset to query CrossRef. Skipping CrossRef augmentation.")
    df_main['Fetched_References_PMIDs_Combined'] = df_main['Fetched_References_PMIDs_EFetch'] # Fallback

# Clean up intermediate EFetch ref column if desired, or keep for comparison
# df_main.drop(columns=['Fetched_References_PMIDs_EFetch'], inplace=True, errors='ignore')

print("\nSample of data after CrossRef reference augmentation:")
display(df_main[['PMID', 'Fetched_DOI', 'Fetched_References_PMIDs_Combined']].head())
# Verify: Check if Fetched_References_PMIDs_Combined has more entries than EFetch only
if 'Fetched_References_PMIDs_EFetch' in df_main.columns:
    ref_count_efetch = df_main['Fetched_References_PMIDs_EFetch'].apply(len).sum()
    ref_count_combined = df_main['Fetched_References_PMIDs_Combined'].apply(len).sum()
    print(f"Total references from EFetch only: {ref_count_efetch}")
    print(f"Total references after CrossRef augmentation: {ref_count_combined}")

print("Cell 4: CrossRef Reference Augmentation Complete.")


# =============================================================================
# CELL 5: Text Preparation for Semantic Similarity & TF-IDF Keywords
# =============================================================================
print("\nPreparing text variations for semantic analysis...")

def preprocess_text_for_nlp(text):
    if not isinstance(text, str): return ""
    text = text.lower()

    # 1. Remove script and style elements, including their content
    #    The pattern [^>]* matches any attributes within the opening tag.
    #    The pattern .*? is a non-greedy match for the content between tags.
    #    re.DOTALL ensures that . matches newline characters as well.
    #    re.IGNORECASE makes the tag matching case-insensitive.
    text = re.sub(r'<script[^>]*>.*?</script>', '', text, flags=re.DOTALL | re.IGNORECASE)
    text = re.sub(r'<style[^>]*>.*?</style>', '', text, flags=re.DOTALL | re.IGNORECASE)
    
    # 2. Remove any remaining HTML tags (e.g., <p>, <div>, <span>, <html> itself)
    #    This will remove the tags but leave their inner content if not handled above.
    text = re.sub(r'<[^>]+>', '', text)
    
    # 3. Filter unwanted characters: Keep alphanumeric, whitespace, dot (.), hyphen (-)
    #    This will remove things like !, ?, quotes, etc.
    text = re.sub(r'[^\w\s\.\-]', '', text)
    
    # 4. Normalize whitespace: Replace multiple spaces/newlines with a single space and strip leading/trailing spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# --- Define functions to create different text combinations ---
# This is where "Combined_Text" and its variations are explicitly defined.

def create_text_from_fields(row, field_weights):
    """
    Creates a combined text string from specified fields in a row, applying weights.
    A field's text is repeated 'ceil(weight)' times if weight > 0.
    More robust handling of potential Series/array content in cells.
    """
    texts_to_combine = []
    for field, weight in field_weights.items():
        if field not in row:
            # print(f"Debug: Field '{field}' not in row. Skipping.") # Optional debug
            continue

        content_val = row[field] # Get the content value

        # If content_val is a pandas Series or numpy array, we need to handle it.
        # This usually means we want to operate on its elements if it's meant to be a list,
        # or take the first element if it was unexpectedly a single-element array.
        # For this function's purpose, if a field is unexpectedly a Series,
        # we'll try to treat it as a list of its values or a single string.
        
        actual_content_for_processing = []
        is_list_like_input = False

        if isinstance(content_val, (pd.Series, np.ndarray)):
            # If it's a Series/array, convert to list. This might happen if a cell contained a list-like string
            # that pandas tried to interpret, or if data was assigned incorrectly.
            # We are assuming if it's an array, it's intended to be a list of items for that field.
            actual_content_for_processing = [item for item in content_val if pd.notna(item)]
            is_list_like_input = True
            # print(f"Debug: Field '{field}' was Series/array, converted to list: {actual_content_for_processing}") # Optional debug
        elif isinstance(content_val, list):
            actual_content_for_processing = [item for item in content_val if pd.notna(item)]
            is_list_like_input = True
        elif pd.notna(content_val): # Scalar value
            actual_content_for_processing = [str(content_val)] # Treat as a list with one string item
            is_list_like_input = False # Though processed as a list, its origin was scalar
        else: # Content is NaN or similar
            # print(f"Debug: Field '{field}' is NaN. Skipping.") # Optional debug
            continue # Skip this field if it's NaN

        # Now actual_content_for_processing is always a list of non-NaN items (or an empty list)
        
        if weight > 0 and actual_content_for_processing: # Process if weight is positive and there's content
            if is_list_like_input: # If original was list-like (MeSH, etc.)
                content_str = " ".join(map(str, actual_content_for_processing))
            else: # If original was scalar, actual_content_for_processing has one item
                content_str = str(actual_content_for_processing[0])
            
            preprocessed_content_str = preprocess_text_for_nlp(content_str)

            if preprocessed_content_str:
                repeat_count = int(np.ceil(weight))
                texts_to_combine.extend([preprocessed_content_str] * repeat_count)
                
    return " ".join(texts_to_combine)

# Define different text combination strategies
TEXT_STRATEGIES = {
    'TitleAbstract': {'Fetched_Title': 1, 'Fetched_Abstract': 1},
    'TitleAbstractMeSH': {'Fetched_Title': 1, 'Fetched_Abstract': 1, 'Fetched_MeSH_Terms': 1},
    'TitleEmphasis': {'Fetched_Title': 3, 'Fetched_Abstract': 1}, # Title weighted 3x
    'AbstractOnly': {'Fetched_Abstract': 1},
    # Add more strategies as needed, e.g., including original 'Study Aim/Focus' if fetched abstract is poor
    'FullInputText': { # This uses original excel input if fetched is missing
        'Title': 1, 'Study Aim/Focus': 1,
        'P: Population Details (Age, N, Condition/Surgery Type, Key Characteristics)': 0.5,
        'I: Intervention(s) Details (ERAS Protocol Elements, Anesthetic Techniques, eHealth/App Use, Specific Drugs/Dosages, Other Interventions)': 0.5,
        'O: Key Outcomes Measured & Findings (LOS, Pain Scores, Opioid Use, Complications, Functional Recovery, Adherence, Patient Satisfaction, Cost, Specific Findings Relevant to Proposal)': 0.5
    }
}

# Apply these strategies to create new columns
for strategy_name, field_cfg in TEXT_STRATEGIES.items():
    col_name = f"Text_{strategy_name}"
    df_main[col_name] = df_main.apply(lambda row: create_text_from_fields(row, field_cfg), axis=1)
    print(f"  Created text column: {col_name}")
    display(df_main[[col_name]].head(1)) # Show an example

# --- TF-IDF Keyword Extraction ---
# We'll use 'Text_TitleAbstract' as the source for keyword extraction for simplicity
KEYWORD_SOURCE_TEXT_COL = 'Text_TitleAbstract'
if KEYWORD_SOURCE_TEXT_COL not in df_main.columns:
    print(f"Warning: {KEYWORD_SOURCE_TEXT_COL} not found for TF-IDF keyword extraction. Skipping.")
    df_main['TFIDF_Keywords'] = [[] for _ in range(len(df_main))]
else:
    print(f"\nExtracting TF-IDF keywords from '{KEYWORD_SOURCE_TEXT_COL}'...")
    tfidf_vectorizer = TfidfVectorizer(stop_words=STOP_WORDS, max_features=5000, ngram_range=(1,2))
    try:
        tfidf_matrix = tfidf_vectorizer.fit_transform(df_main[KEYWORD_SOURCE_TEXT_COL].fillna(""))
        feature_names = tfidf_vectorizer.get_feature_names_out()
        
        def get_top_tfidf_keywords(row_vector, n_top=NUM_KEYWORDS_TFIDF):
            # Sort by TF-IDF score and get top N
            sorted_indices = np.argsort(row_vector.toarray().flatten())[::-1]
            top_keywords = [feature_names[i] for i in sorted_indices[:n_top] if row_vector[0, i] > 0] # Only if score > 0
            return top_keywords

        df_main['TFIDF_Keywords'] = [get_top_tfidf_keywords(tfidf_matrix[i]) for i in range(tfidf_matrix.shape[0])]
        print("  TF-IDF keywords extracted.")
        display(df_main[['PMID', 'TFIDF_Keywords']].head(2))
    except ValueError as e:
        print(f"  Error during TF-IDF vectorization (e.g., all documents are empty or stop words): {e}. Skipping keyword extraction.")
        df_main['TFIDF_Keywords'] = [[] for _ in range(len(df_main))]


# Preprocess the PROPOSAL_QUERY_TEXT as well
PROCESSED_PROPOSAL_QUERY_TEXT = preprocess_text_for_nlp(PROPOSAL_QUERY_TEXT)
# Extract TF-IDF keywords for the proposal query text
if KEYWORD_SOURCE_TEXT_COL in df_main.columns and 'tfidf_vectorizer' in locals(): # if vectorizer was fit
    try:
        proposal_tfidf_vector = tfidf_vectorizer.transform([PROCESSED_PROPOSAL_QUERY_TEXT])
        PROPOSAL_TFIDF_KEYWORDS = get_top_tfidf_keywords(proposal_tfidf_vector[0])
        print(f"\nTF-IDF Keywords for PROPOSAL_QUERY_TEXT: {PROPOSAL_TFIDF_KEYWORDS}")
    except ValueError:
        print("Could not extract TF-IDF keywords for proposal (e.g., proposal text too short after preprocessing).")
        PROPOSAL_TFIDF_KEYWORDS = []
else:
    PROPOSAL_TFIDF_KEYWORDS = []
    print("TF-IDF vectorizer not available for proposal keyword extraction.")


print("Cell 5: Text Preparation Complete.")

# =============================================================================
# CELL 6: Semantic Similarity Metrics Calculation
# =============================================================================
print("\nCalculating Semantic Similarity Scores...")
try:
    semantic_model = SentenceTransformer(SEMANTIC_MODEL_NAME)
    print(f"  Loaded SentenceTransformer model: {SEMANTIC_MODEL_NAME}")

    # --- 1. Semantic Similarity on different text combinations ---
    proposal_embedding_main = semantic_model.encode(PROCESSED_PROPOSAL_QUERY_TEXT, convert_to_tensor=True)

    for strategy_name in TEXT_STRATEGIES.keys():
        text_col = f"Text_{strategy_name}"
        score_col = f"SemSim_{strategy_name}"
        print(f"  Calculating semantic similarity for strategy: {strategy_name} (using column '{text_col}')...")
        
        # Filter out rows where text_col might be empty after processing
        valid_texts_series = df_main[text_col][df_main[text_col].str.strip() != ""]
        if valid_texts_series.empty:
            print(f"    No valid text found for strategy {strategy_name}. Skipping.")
            df_main[score_col] = np.nan
            continue

        doc_embeddings = semantic_model.encode(valid_texts_series.tolist(), convert_to_tensor=True, show_progress_bar=False)
        cosine_scores = util.cos_sim(proposal_embedding_main, doc_embeddings)
        
        # Assign scores back, handling cases where some rows had no text
        scores_to_assign = np.full(len(df_main), np.nan)
        valid_indices = valid_texts_series.index
        scores_to_assign[valid_indices] = cosine_scores[0].cpu().numpy()
        df_main[score_col] = scores_to_assign
        print(f"    {score_col} calculated.")

    # --- 2. Semantic Similarity based on TF-IDF Keywords ---
    # This compares the average embedding of a document's keywords to the average embedding of the proposal's keywords.
    # More sophisticated methods exist (e.g., Word Mover's Distance on keyword embeddings) but are complex.
    if PROPOSAL_TFIDF_KEYWORDS and 'TFIDF_Keywords' in df_main.columns:
        print("  Calculating semantic similarity based on TF-IDF Keywords...")
        # Get embedding for proposal's keywords
        if PROPOSAL_TFIDF_KEYWORDS: # Ensure there are keywords
            proposal_keyword_embeddings = semantic_model.encode(PROPOSAL_TFIDF_KEYWORDS, convert_to_tensor=True)
            avg_proposal_keyword_embedding = torch.mean(proposal_keyword_embeddings, dim=0, keepdim=True) # PyTorch tensor
        else: # Handle case where proposal has no keywords
            avg_proposal_keyword_embedding = None
            print("    Proposal has no TF-IDF keywords. Skipping keyword similarity.")

        def calculate_keyword_similarity(doc_keywords_list):
            if not doc_keywords_list or avg_proposal_keyword_embedding is None: return np.nan
            doc_keyword_embeddings = semantic_model.encode(doc_keywords_list, convert_to_tensor=True)
            if doc_keyword_embeddings.nelement() == 0: return np.nan # No valid embeddings for doc keywords
            avg_doc_keyword_embedding = torch.mean(doc_keyword_embeddings, dim=0, keepdim=True)
            
            # Cosine similarity between the two average keyword embeddings
            return util.cos_sim(avg_proposal_keyword_embedding, avg_doc_keyword_embedding)[0,0].item()

        df_main['SemSim_TFIDF_Keywords'] = df_main['TFIDF_Keywords'].apply(calculate_keyword_similarity)
        print("    SemSim_TFIDF_Keywords calculated.")
    else:
        print("  Skipping TF-IDF Keyword-based semantic similarity (no proposal keywords or document keywords).")
        df_main['SemSim_TFIDF_Keywords'] = np.nan

except Exception as e:
    print(f"ERROR during Semantic Similarity calculation: {e}")
    # Set all semantic sim columns to NaN if model fails to load or run
    for strategy_name in TEXT_STRATEGIES.keys(): df_main[f"SemSim_{strategy_name}"] = np.nan
    df_main['SemSim_TFIDF_Keywords'] = np.nan

display(df_main[[col for col in df_main.columns if "SemSim_" in col]].head())
print("Cell 6: Semantic Similarity Metrics Complete.")

# =============================================================================
# CELL 7: Naive Bayes Classifier
# =============================================================================
print("\nAttempting Naive Bayes Classification for Relevance...")
# Use 'Text_TitleAbstractMeSH' as the text input for Naive Bayes for good coverage
NB_TEXT_INPUT_COL = 'Text_TitleAbstractMeSH'
NB_TARGET_COL = 'Relevance Score (1-10)' # Ground truth

if NB_TEXT_INPUT_COL not in df_main.columns:
    print(f"  Input column '{NB_TEXT_INPUT_COL}' for Naive Bayes not found. Skipping.")
    df_main['NB_High_Relevance_Prob'] = np.nan
elif NB_TARGET_COL not in df_main.columns:
    print(f"  Target column '{NB_TARGET_COL}' for Naive Bayes not found. Skipping.")
    df_main['NB_High_Relevance_Prob'] = np.nan
else:
    # Bin relevance scores for classification
    df_main['Relevance_Bin'] = pd.cut(df_main[NB_TARGET_COL],
                                      bins=RELEVANCE_BINS,
                                      labels=RELEVANCE_LABELS,
                                      right=True, include_lowest=True)

    # Check for class diversity
    value_counts = df_main['Relevance_Bin'].value_counts()
    valid_classes = value_counts[value_counts >= MIN_SAMPLES_PER_CLASS_NB]

    if len(valid_classes) < 2:
        print(f"  Skipping Naive Bayes: Insufficient class diversity or samples per class (need at least 2 classes with >= {MIN_SAMPLES_PER_CLASS_NB} samples). Found: {value_counts.to_dict()}")
        df_main['NB_High_Relevance_Prob'] = np.nan
    else:
        print(f"  Sufficient class diversity for Naive Bayes. Training model using '{NB_TEXT_INPUT_COL}'. Bins: {value_counts.to_dict()}")
        df_train_nb = df_main.dropna(subset=[NB_TEXT_INPUT_COL, 'Relevance_Bin'])
        
        if len(df_train_nb) < (MIN_SAMPLES_PER_CLASS_NB * 2): # Basic check for enough total data
             print(f"  Too few overall samples ({len(df_train_nb)}) after dropping NaNs for Naive Bayes. Skipping.")
             df_main['NB_High_Relevance_Prob'] = np.nan
        else:
            vectorizer_nb = TfidfVectorizer(stop_words=STOP_WORDS, max_features=5000)
            X_nb = vectorizer_nb.fit_transform(df_train_nb[NB_TEXT_INPUT_COL])
            y_nb = df_train_nb['Relevance_Bin']
            
            try:
                nb_model = MultinomialNB()
                nb_model.fit(X_nb, y_nb)
                
                # Predict probabilities on the whole dataset (where text is available)
                # Rows with no text for NB_TEXT_INPUT_COL will get NaN
                probs_nb = np.full((len(df_main), len(nb_model.classes_)), np.nan)
                
                # Identify rows with valid text to predict on
                valid_text_for_nb_indices = df_main[df_main[NB_TEXT_INPUT_COL].str.strip() != ""].index
                if not valid_text_for_nb_indices.empty:
                    X_pred_nb = vectorizer_nb.transform(df_main.loc[valid_text_for_nb_indices, NB_TEXT_INPUT_COL])
                    predicted_probs_for_valid = nb_model.predict_proba(X_pred_nb)
                    
                    # Place predicted probabilities into the correct rows of probs_nb
                    # Create a temporary DataFrame for easier assignment
                    temp_probs_df = pd.DataFrame(predicted_probs_for_valid, index=valid_text_for_nb_indices, columns=nb_model.classes_)
                    
                    if 'High' in nb_model.classes_:
                        df_main['NB_High_Relevance_Prob'] = temp_probs_df['High'] # This will align by index
                    else:
                        print("    Warning: 'High' relevance class not found in Naive Bayes model. Setting probabilities to 0.")
                        df_main['NB_High_Relevance_Prob'] = 0.0
                else: # No valid text in the entire dataset for prediction
                    df_main['NB_High_Relevance_Prob'] = np.nan

                print("  Naive Bayes probabilities calculated.")

            except ValueError as ve: # e.g., if a class has too few samples during fit after internal splits
                print(f"    Skipping Naive Bayes due to ValueError during model training/prediction: {ve}")
                df_main['NB_High_Relevance_Prob'] = np.nan
                
# Clean up Relevance_Bin if not needed further, or keep for inspection
# df_main.drop(columns=['Relevance_Bin'], inplace=True, errors='ignore')
print("Cell 7: Naive Bayes Classification Complete.")

# =============================================================================
# CELL 8: Structural Metrics Calculation
# =============================================================================
print("\nCalculating Structural Metrics (Co-citation, Bibliographic Coupling)...")

# --- Define Core Set based on relevance for structural metrics ---
df_core_papers_struct = df_main[df_main[NB_TARGET_COL] >= CORE_RELEVANCE_THRESHOLD_FOR_STRUCTURAL]
core_set_pmids_struct = df_core_papers_struct['PMID'].tolist()
if core_set_pmids_struct:
    print(f"  Core set for structural metrics defined: {len(core_set_pmids_struct)} PMIDs (Relevance >= {CORE_RELEVANCE_THRESHOLD_FOR_STRUCTURAL}).")
else:
    print("  Warning: No core set defined for structural metrics (no papers met threshold). Scores will be 0 or NaN.")

# Create PMID to PubDate mapping for papers in our dataset (needed for age attenuation)
pmid_to_date_map_struct = df_main.dropna(subset=['Fetched_PubDate', 'PMID']).set_index('PMID')['Fetched_PubDate'].to_dict()

# --- Modified Co-citation Score with Core Set ---
def calculate_modified_cocitation(target_pmid, target_cited_by_pmids, core_pmids_dict, pmid_date_map):
    if not core_pmids_dict or target_pmid in core_pmids_dict: return 0.0
    if not target_cited_by_pmids: return 0.0

    target_cited_by_set = set(target_cited_by_pmids)
    total_modified_cocit_score = 0.0
    
    for core_pmid, core_cited_by_pmids in core_pmids_dict.items():
        if target_pmid == core_pmid: continue
        core_cited_by_set = set(core_cited_by_pmids)
        common_citers = target_cited_by_set.intersection(core_cited_by_set)
        
        pairwise_cocit_score = 0.0
        for citer_pmid in common_citers:
            date_citer = pmid_date_map.get(citer_pmid)
            date_target = pmid_date_map.get(target_pmid)
            date_core = pmid_date_map.get(core_pmid)

            if date_citer and date_target and date_core:
                try:
                    age_target_at_citation = (date_citer - date_target).days / 365.25
                    age_core_at_citation = (date_citer - date_core).days / 365.25
                    if age_target_at_citation >= 0 and age_core_at_citation >= 0: # Citer must be published after/on same day
                        # Attenuation: add 1 to avoid division by zero, smaller age = higher weight
                        weight = 1.0 / ((age_target_at_citation + 1) * (age_core_at_citation + 1))
                        pairwise_cocit_score += weight
                except TypeError: pass # Should not happen with valid dates
        total_modified_cocit_score += pairwise_cocit_score
    return total_modified_cocit_score

# Prepare core_pmids_dict for co-citation: {pmid: [cited_by_list]}
core_pmids_cocit_dict = {}
if core_set_pmids_struct:
    for _, row_core in df_core_papers_struct.iterrows():
        core_pmids_cocit_dict[row_core['PMID']] = row_core['Fetched_Cited_By_PMIDs']

df_main['Struct_Mod_CoCite'] = df_main.apply(
    lambda row: calculate_modified_cocitation(
        row['PMID'], row['Fetched_Cited_By_PMIDs'], core_pmids_cocit_dict, pmid_to_date_map_struct
    ), axis=1
)
print("  Modified Co-citation scores calculated.")

# --- Bibliographic Coupling Score with Core Set ---
def calculate_bibliographic_coupling(target_pmid, target_refs_pmids, core_pmids_refs_dict):
    if not core_pmids_refs_dict or target_pmid in core_pmids_refs_dict: return 0
    if not target_refs_pmids: return 0
        
    target_refs_set = set(target_refs_pmids)
    total_bib_coupling_score = 0
    for core_pmid, core_refs_pmids in core_pmids_refs_dict.items():
        if target_pmid == core_pmid: continue
        core_refs_set = set(core_refs_pmids)
        common_refs = target_refs_set.intersection(core_refs_set)
        total_bib_coupling_score += len(common_refs)
    return total_bib_coupling_score

# Prepare core_pmids_dict for bibliographic coupling: {pmid: [references_list]}
core_pmids_bibcoup_dict = {}
if core_set_pmids_struct:
     for _, row_core in df_core_papers_struct.iterrows():
        core_pmids_bibcoup_dict[row_core['PMID']] = row_core['Fetched_References_PMIDs_Combined']

df_main['Struct_Bib_Couple'] = df_main.apply(
    lambda row: calculate_bibliographic_coupling(
        row['PMID'], row['Fetched_References_PMIDs_Combined'], core_pmids_bibcoup_dict
    ), axis=1
)
print("  Bibliographic Coupling scores calculated.")
display(df_main[['PMID', 'Struct_Mod_CoCite', 'Struct_Bib_Couple']].head())
print("Cell 8: Structural Metrics Calculation Complete.")


# =============================================================================
# CELL 9: Benchmarking & Correlation Analysis
# =============================================================================
print("\n--- Benchmarking and Correlation Analysis ---")
ground_truth_col = 'Relevance Score (1-10)'

# Identify all calculated metric columns
metric_cols_for_corr = [col for col in df_main.columns if "SemSim_" in col or "NB_" in col or "Struct_" in col]

if not metric_cols_for_corr:
    print("No metric columns found for correlation. Skipping.")
else:
    # Ensure ground truth column exists
    if ground_truth_col not in df_main.columns:
        print(f"Ground truth column '{ground_truth_col}' not found. Skipping correlation.")
    else:
        df_for_corr = df_main[[ground_truth_col] + metric_cols_for_corr].copy()
        # Convert all to numeric, coercing errors for cases where some scores might be non-numeric (shouldn't happen)
        for col in df_for_corr.columns:
            df_for_corr[col] = pd.to_numeric(df_for_corr[col], errors='coerce')

        # Drop rows where ground_truth might be NaN, or any metric is NaN for pair-wise correlation
        # For overall correlation table, better to keep NaNs and let .corr() handle pair-wise.
        
        correlations = df_for_corr.corr(method='spearman')
        
        print(f"\nSpearman Correlation with '{ground_truth_col}':")
        # Exclude self-correlation, sort by ground_truth_col
        ground_truth_corrs = correlations[[ground_truth_col]].drop(ground_truth_col).sort_values(by=ground_truth_col, ascending=False)
        display(ground_truth_corrs)

        plt.figure(figsize=(max(8, len(metric_cols_for_corr)*0.5) , max(6, len(metric_cols_for_corr)*0.4))) # Dynamic size
        sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
        plt.title(f'Spearman Correlation Heatmap of Metrics (vs. {ground_truth_col})')
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

print("Cell 9: Benchmarking and Correlation Analysis Complete.")

# =============================================================================
# CELL 10: Plotting Individual Metric Scatter Plots
# =============================================================================
print("\n--- Plotting Individual Metric Scatter Plots ---")
if not metric_cols_for_corr:
    print("No metric columns found for plotting. Skipping.")
elif ground_truth_col not in df_main.columns:
    print(f"Ground truth column '{ground_truth_col}' not found. Skipping plotting.")
else:
    num_metrics_plot = len(metric_cols_for_corr)
    if num_metrics_plot > 0:
        # Determine grid size for subplots
        cols_subplot = 2
        rows_subplot = (num_metrics_plot + cols_subplot - 1) // cols_subplot
        
        fig, axes = plt.subplots(nrows=rows_subplot, ncols=cols_subplot, figsize=(15, 5 * rows_subplot))
        axes = np.array(axes).flatten() # Flatten to 1D array for easy iteration, handle single plot case

        for i, metric in enumerate(metric_cols_for_corr):
            ax = axes[i]
            if metric in df_main.columns and df_main[metric].notna().any():
                # Jitter x-values slightly if they are very discrete to see overlapping points
                # sns.stripplot(data=df_main, x=metric, y=ground_truth_col, ax=ax, alpha=0.6, jitter=0.1)
                sns.scatterplot(data=df_main, x=metric, y=ground_truth_col, ax=ax, alpha=0.7)
                ax.set_title(f'{metric} vs. {ground_truth_col}')
                ax.set_xlabel(metric)
                ax.set_ylabel(ground_truth_col)
            else:
                ax.set_title(f'{metric}\n(No data or all NaN)')
                ax.axis('off')
        
        # Hide any unused subplots
        for j in range(i + 1, len(axes)):
            fig.delaxes(axes[j])

        plt.tight_layout()
        plt.show()
    else:
        print("No valid metrics available for scatter plotting.")
print("Cell 10: Plotting Complete.")

# =============================================================================
# CELL 11: Display Final DataFrame Sample
# =============================================================================
print("\n--- Final DataFrame Sample with All Calculated Scores ---")
# Select key columns for display, including original, some fetched, and all new scores
display_cols_final = ['PMID', 'Title', ground_truth_col] + \
                     ['Fetched_Title', 'Fetched_DOI', 'Fetched_PubDate'] + \
                     metric_cols_for_corr + \
                     ['Fetched_References_PMIDs_Combined', 'Fetched_Cited_By_PMIDs']


# Ensure all columns exist before trying to display them
final_cols_to_show = [col for col in display_cols_final if col in df_main.columns]
display(df_main[final_cols_to_show].head())

# Save to new Excel if needed
# output_excel_path = "screening_ERAS_augmented_and_scored.xlsx"
# df_main.to_excel(output_excel_path, index=False)
# print(f"\nFull augmented DataFrame saved to: {output_excel_path}")

print("\n--- Analysis Script Fully Complete ---")

# Import Pytorch for SemSim_TFIDF_Keywords if not already (it's used by sentence-transformers)
import torch
```

```python
# =============================================================================
# CELL 1: Setup, Imports, and Configuration
# =============================================================================
import requests
import time
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
import datetime
from urllib.parse import quote
from tqdm.notebook import tqdm
import re
import json
import torch # For sentence_transformers and direct use in SemSim_TFIDF_Keywords

# Text Processing and ML
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sentence_transformers import SentenceTransformer, util
import scipy.stats # For Spearman correlation p-values

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Display (though we'll use print with .to_string() for DataFrames)
from IPython.display import HTML # HTML might still be used by tqdm or other libraries implicitly

# --- Configuration ---
XLSX_FILE_PATH = "C:/Users/Galaxy/Downloads/screening_ERAS.xlsx" # !!! UPDATE THIS PATH !!!
NCBI_API_KEY   = "YOUR_NCBI_API_KEY"  # !!! REPLACE with your NCBI API key !!!
CROSSREF_MAILTO= "your_email@example.com" # !!! REPLACE with your email for CrossRef polite pool !!!

# --- NCBI/CrossRef Batch sizes and Delays ---
EFETCH_BATCH_SIZE = 100
ELINK_BATCH_SIZE  = 100
ID_CONVERTER_BATCH_SIZE_CR = 50
CROSSREF_FILTER_BATCH_SIZE = 30
API_DELAY = 0.2 if NCBI_API_KEY and NCBI_API_KEY != "YOUR_NCBI_API_KEY" else 0.4

# --- Semantic Similarity Model ---
SEMANTIC_MODEL_NAME = 'all-MiniLM-L6-v2'

# --- User-Defined Proposal Query Text ---
PROPOSAL_QUERY_TEXT = """
Developing and validating an ERAS (Enhanced Recovery After Surgery) protocol
for pediatric airway surgery. This involves creating a mobile application for collecting
anthropometric, clinical, and anesthetic data to inform the ERAS protocol.
The focus is on optimizing perioperative management, including anesthetic techniques,
for different age groups and comorbidity profiles to ensure shorter, more comfortable
post-surgical recovery, reduce adverse events like bronchospasm and desaturation,
and improve overall quality of care and patient safety in procedures such as
tracheoscopies, tracheal dilations, bronchoscopies, and tracheoplasties.
The aim is to generate an evidence-based, app-driven ERAS protocol that can serve
as a model for similar pediatric airway services.
"""

# --- Naive Bayes Configuration ---
RELEVANCE_BINS = [0, 3, 7, 10]
RELEVANCE_LABELS = ['Low', 'Medium', 'High']
MIN_SAMPLES_PER_CLASS_NB = 5

# --- TF-IDF Keyword Extraction Configuration ---
NUM_KEYWORDS_TFIDF = 10

# --- Core Set Definition for Structural Metrics ---
CORE_RELEVANCE_THRESHOLD_FOR_STRUCTURAL = 9

# --- Setup NLTK Stopwords ---
try:
    STOP_WORDS = stopwords.words('english')
except LookupError:
    print("NLTK stopwords not found. Downloading...")
    nltk.download('stopwords')
    STOP_WORDS = stopwords.words('english')

# --- Requests Session ---
session = requests.Session()
adapter = requests.adapters.HTTPAdapter(pool_connections=50, pool_maxsize=50, max_retries=3)
session.mount("https://", adapter)
session.mount("http://", adapter)

session_cr = requests.Session()
adapter_cr = requests.adapters.HTTPAdapter(pool_connections=10, pool_maxsize=10, max_retries=3)
session_cr.mount("https://", adapter_cr)

# --- Pandas Display Options for to_string() ---
# pd.set_option('display.max_rows', 100)
# pd.set_option('display.max_columns', 20)
# pd.set_option('display.width', 1000) # Adjust for wider output if needed

print("Cell 1: Setup and Configuration Complete.")
print(f"PROPOSAL_QUERY_TEXT set to: '{PROPOSAL_QUERY_TEXT[:100]}...'")

# =============================================================================
# CELL 2: NCBI & CrossRef API Helper Functions
# =============================================================================
def _get_ncbi_params(extra=None, tool_name="eras_relevance_v2_txt"):
    params = {"tool": tool_name, "email": CROSSREF_MAILTO}
    if NCBI_API_KEY and NCBI_API_KEY != "YOUR_NCBI_API_KEY":
        params["api_key"] = NCBI_API_KEY
    if extra: params.update(extra)
    return params

def parse_efetch_article_xml(pubmed_article_node):
    pmid_node = pubmed_article_node.find(".//MedlineCitation/PMID")
    pmid = pmid_node.text if pmid_node is not None else None
    if not pmid: return None
    article_node = pubmed_article_node.find(".//MedlineCitation/Article")
    if article_node is None: return {'PMID_Fetched': pmid}
    title = article_node.findtext(".//ArticleTitle", default="")
    abstract_parts = [at.text for at in article_node.findall(".//Abstract/AbstractText") if at.text]
    abstract = "\n".join(abstract_parts) if abstract_parts else ""
    mesh_headings = []
    for mesh_node in pubmed_article_node.findall(".//MedlineCitation/MeshHeadingList/MeshHeading"):
        desc_name = mesh_node.findtext(".//DescriptorName")
        qual_names = [q.text for q in mesh_node.findall(".//QualifierName")]
        full_mesh = desc_name
        if qual_names: full_mesh += "/" + "/".join(qual_names)
        if desc_name: mesh_headings.append(full_mesh)
    pub_date = None
    MONTH_MAP = {m: i for i, m in enumerate(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], 1)}
    date_priority = [".//PubmedData/History/PubMedPubDate[@PubStatus='pubmed']", ".//PubmedData/History/PubMedPubDate[@PubStatus='entrez']", ".//ArticleDate", ".//Journal/JournalIssue/PubDate"]
    for path in date_priority:
        date_node = pubmed_article_node.find(path)
        if date_node is not None:
            year, month, day = date_node.findtext("Year"), date_node.findtext("Month"), date_node.findtext("Day")
            try:
                m_val = MONTH_MAP.get(month, int(month) if month and month.isdigit() else 1)
                pub_date = datetime.date(int(year), m_val, int(day)); break
            except: continue
    doi = ""
    doi_node = pubmed_article_node.find(f".//PubmedData/ArticleIdList/ArticleId[@IdType='doi']")
    if doi_node is not None and doi_node.text: doi = doi_node.text
    else:
        for art_id in article_node.findall(".//ELocationID[@EIdType='doi']"):
            if art_id.text: doi = art_id.text; break
    journal_title = article_node.findtext(".//Journal/Title", default="")
    efetch_references_pmids = [ref_node.find(".//ArticleIdList/ArticleId[@IdType='pubmed']").text for ref_node in pubmed_article_node.findall(".//ReferenceList/Reference") if ref_node.find(".//ArticleIdList/ArticleId[@IdType='pubmed']") is not None and ref_node.find(".//ArticleIdList/ArticleId[@IdType='pubmed']").text]
    return {"PMID_Fetched": pmid, "Fetched_Title": title, "Fetched_Abstract": abstract, "Fetched_MeSH_Terms": mesh_headings, "Fetched_PubDate": pub_date, "Fetched_DOI": doi, "Fetched_Journal": journal_title, "Fetched_References_PMIDs_EFetch": efetch_references_pmids}

def fetch_pubmed_details_batch(pmids_list):
    all_data = []
    pmids_set = set(pmids_list)
    print(f"Fetching PubMed details for {len(pmids_list)} PMIDs (EFetch)...")
    for i in tqdm(range(0, len(pmids_list), EFETCH_BATCH_SIZE), desc="PubMed EFetch Batches"):
        batch_pmids = pmids_list[i:i + EFETCH_BATCH_SIZE]
        params = _get_ncbi_params({"db": "pubmed", "retmode": "xml", "id": ",".join(batch_pmids)})
        try:
            r = session.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi", data=params, timeout=180)
            r.raise_for_status(); root = ET.fromstring(r.content)
            for article_xml in root.findall(".//PubmedArticle"):
                parsed_data = parse_efetch_article_xml(article_xml)
                if parsed_data and parsed_data['PMID_Fetched'] in pmids_set: all_data.append(parsed_data)
        except Exception as e: print(f"  - ERROR EFetch batch: {e}")
        time.sleep(API_DELAY)
    df = pd.DataFrame(all_data)
    if not df.empty: df['Fetched_PubDate'] = pd.to_datetime(df['Fetched_PubDate'], errors='coerce').dt.date
    return df

def fetch_pubmed_citations_batch(pmids_list):
    citations_map = {pmid: [] for pmid in pmids_list}
    print(f"Fetching PubMed citations for {len(pmids_list)} PMIDs (ELink)...")
    for i in tqdm(range(0, len(pmids_list), ELINK_BATCH_SIZE), desc="PubMed ELink Batches"):
        batch_pmids = pmids_list[i:i + ELINK_BATCH_SIZE]
        params = _get_ncbi_params({"dbfrom": "pubmed", "db": "pubmed", "linkname": "pubmed_pubmed_citedin", "id": ",".join(batch_pmids)})
        try:
            r = session.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/elink.fcgi", data=params, timeout=120)
            r.raise_for_status(); root = ET.fromstring(r.content)
            for linkset_node in root.findall(".//LinkSet"):
                source_pmid = linkset_node.findtext("./IdList/Id")
                if source_pmid in citations_map:
                    for link_node in linkset_node.findall(".//LinkSetDb/Link/Id"):
                        if link_node.text: citations_map[source_pmid].append(link_node.text)
        except Exception as e: print(f"  - ERROR ELink batch: {e}")
        time.sleep(API_DELAY)
    return citations_map

def get_crossref_refs_via_filter_batch(doi_list):
    if not doi_list: return {}
    print(f"  Fetching reference DOIs from CrossRef for {len(doi_list)} publication DOIs...")
    all_results = {}
    for i in tqdm(range(0, len(doi_list), CROSSREF_FILTER_BATCH_SIZE), desc="CrossRef Ref Batches", leave=False):
        batch_dois = doi_list[i:i + CROSSREF_FILTER_BATCH_SIZE]
        filter_value = ",".join([f"doi:{d.lower()}" for d in batch_dois])
        params = {"filter": filter_value, "rows": len(batch_dois), "select": "DOI,reference"}
        headers = {"User-Agent": f"RelevanceAnalyzer/1.0 (mailto:{CROSSREF_MAILTO})"}
        base_url = "https://api.crossref.org/works"
        try:
            req = requests.Request('GET', base_url, params=params).prepare()
            if len(req.url) > 4000: print(f"    WARN: URL for CrossRef batch might be too long ({len(req.url)} chars).")
            r = session_cr.get(base_url, params=params, headers=headers, timeout=120)
            r.raise_for_status(); data = r.json().get("message", {}).get("items", [])
            for item in data:
                item_doi_cr = item.get("DOI")
                matched_input_doi = next((d_in for d_in in batch_dois if d_in.lower() == item_doi_cr.lower()), None)
                if matched_input_doi:
                    refs = item.get("reference", [])
                    ref_dois = [ref.get("DOI") for ref in refs if ref.get("DOI")]
                    all_results[matched_input_doi] = ref_dois
        except Exception as e: print(f"    - ERROR CrossRef batch for DOIs {batch_dois[:2]}...: {e}")
        time.sleep(API_DELAY)
    print(f"  CrossRef fetched reference DOIs for {len(all_results)} publications.") # Changed PMIDs to publications
    return all_results

def batch_dois_to_pmids_ncbi(doi_list_to_convert, batch_size=ID_CONVERTER_BATCH_SIZE_CR):
    if not doi_list_to_convert: return {}
    doi_to_pmid_map = {}
    unique_dois = list(set(d.lower() for d in doi_list_to_convert if d))
    print(f"  Converting {len(unique_dois)} unique reference DOIs to PMIDs via NCBI ESearch...")
    for i in tqdm(range(0, len(unique_dois), batch_size), desc="DOI to PMID (ESearch)", leave=False):
        batch_dois_current_sub_batch = unique_dois[i:i+batch_size]
        query_term = " OR ".join([f"{quote(doi_val)}[DOI]" for doi_val in batch_dois_current_sub_batch])
        params = _get_ncbi_params({"db": "pubmed", "term": query_term, "retmax": str(len(batch_dois_current_sub_batch))})
        try:
            r = session.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi", data=params, timeout=120)
            r.raise_for_status(); root = ET.fromstring(r.content)
            pmids_found = [id_node.text for id_node in root.findall(".//IdList/Id")]
            if pmids_found:
                temp_df_details = fetch_pubmed_details_batch(pmids_found) # EFetch to get DOIs for these PMIDs
                if not temp_df_details.empty:
                    for _, row_detail in temp_df_details.iterrows():
                        fetched_doi = row_detail['Fetched_DOI']
                        if fetched_doi and fetched_doi.lower() in batch_dois_current_sub_batch:
                             doi_to_pmid_map[fetched_doi.lower()] = row_detail['PMID_Fetched']
                             original_case_doi = next((d_orig for d_orig in batch_dois_current_sub_batch if d_orig.lower() == fetched_doi.lower()), None)
                             if original_case_doi: doi_to_pmid_map[original_case_doi] = row_detail['PMID_Fetched']
        except Exception as e: print(f"    - ERROR ESearch for DOI to PMID batch: {e}")
        time.sleep(API_DELAY)
    print(f"  NCBI ESearch mapped {len(doi_to_pmid_map)} DOIs to PMIDs.") # Changed converted to mapped
    return doi_to_pmid_map

print("Cell 2: API Helper Functions Defined.")

# =============================================================================
# CELL 3: Load Initial Data & Core PubMed Augmentation
# =============================================================================
print(f"\nLoading data from: {XLSX_FILE_PATH}")
df_main = pd.read_excel(XLSX_FILE_PATH)
df_main['PMID'] = df_main['PMID'].astype(str)
print(f"Loaded {len(df_main)} records from XLSX.")
print(df_main.head(2).to_string()) # CHANGED TO to_string()

pmids_initial_list = df_main['PMID'].unique().tolist()

df_pubmed_details = fetch_pubmed_details_batch(pmids_initial_list)
if not df_pubmed_details.empty:
    df_main = pd.merge(df_main, df_pubmed_details, left_on='PMID', right_on='PMID_Fetched', how='left')
    df_main.drop(columns=['PMID_Fetched'], inplace=True, errors='ignore')
else:
    print("Warning: No details fetched from PubMed EFetch.")
    for col in ["Fetched_Title", "Fetched_Abstract", "Fetched_MeSH_Terms", "Fetched_PubDate", "Fetched_DOI", "Fetched_Journal", "Fetched_References_PMIDs_EFetch"]:
        if col not in df_main.columns: df_main[col] = np.nan if col != "Fetched_MeSH_Terms" else [[] for _ in range(len(df_main))]

citations_map = fetch_pubmed_citations_batch(pmids_initial_list)
df_main['Fetched_Cited_By_PMIDs'] = df_main['PMID'].map(citations_map).fillna('').apply(list)

for col in ['Fetched_MeSH_Terms', 'Fetched_References_PMIDs_EFetch', 'Fetched_Cited_By_PMIDs']:
    if col in df_main.columns: df_main[col] = df_main[col].apply(lambda x: x if isinstance(x, list) else ([] if pd.isna(x) else ([x] if not isinstance(x,list) else []) ))
    else: df_main[col] = [[] for _ in range(len(df_main))]

print("\nSample of data after core PubMed augmentation (details & citations):")
print(df_main[['PMID', 'Title', 'Fetched_Title', 'Fetched_DOI', 'Fetched_References_PMIDs_EFetch', 'Fetched_Cited_By_PMIDs']].head().to_string()) # CHANGED
print("Cell 3: Core PubMed Augmentation Complete.")

# =============================================================================
# CELL 4: Comprehensive Reference Fetching (CrossRef Augmentation)
# =============================================================================
print("\nStarting Comprehensive Reference Fetching via CrossRef...")
dois_for_crossref = df_main[df_main['Fetched_DOI'].notna() & (df_main['Fetched_DOI'] != "")]['Fetched_DOI'].unique().tolist()
if dois_for_crossref:
    pub_doi_to_ref_dois_map = get_crossref_refs_via_filter_batch(dois_for_crossref)
    all_ref_dois_to_convert = set(ref_doi for ref_list in pub_doi_to_ref_dois_map.values() for ref_doi in ref_list)
    ref_doi_to_pmid_map = batch_dois_to_pmids_ncbi(list(all_ref_dois_to_convert)) if all_ref_dois_to_convert else {}
    if not all_ref_dois_to_convert: print("No reference DOIs found from CrossRef to convert to PMIDs.")

    def combine_references(row):
        efetch_refs = set(row['Fetched_References_PMIDs_EFetch'] if isinstance(row['Fetched_References_PMIDs_EFetch'], list) else [])
        crossref_ref_pmids = set()
        pub_doi = row['Fetched_DOI']
        if pd.notna(pub_doi) and pub_doi in pub_doi_to_ref_dois_map:
            for ref_doi in pub_doi_to_ref_dois_map[pub_doi]:
                ref_pmid = ref_doi_to_pmid_map.get(ref_doi.lower()) # Lookup lowercase
                if ref_pmid: crossref_ref_pmids.add(ref_pmid)
        combined_list = list(set(map(str, efetch_refs.union(crossref_ref_pmids))))
        return [r for r in combined_list if r != row['PMID']]
    df_main['Fetched_References_PMIDs_Combined'] = df_main.apply(combine_references, axis=1)
else:
    print("No DOIs found in the dataset to query CrossRef. Skipping CrossRef augmentation.")
    df_main['Fetched_References_PMIDs_Combined'] = df_main['Fetched_References_PMIDs_EFetch']

print("\nSample of data after CrossRef reference augmentation:")
print(df_main[['PMID', 'Fetched_DOI', 'Fetched_References_PMIDs_Combined']].head().to_string()) # CHANGED
if 'Fetched_References_PMIDs_EFetch' in df_main.columns:
    ref_count_efetch = df_main['Fetched_References_PMIDs_EFetch'].apply(len).sum()
    ref_count_combined = df_main['Fetched_References_PMIDs_Combined'].apply(len).sum()
    print(f"Total references from EFetch only: {ref_count_efetch}")
    print(f"Total references after CrossRef augmentation: {ref_count_combined}")
print("Cell 4: CrossRef Reference Augmentation Complete.")

# =============================================================================
# CELL 5: Text Preparation for Semantic Similarity & TF-IDF Keywords
# =============================================================================
print("\nPreparing text variations for semantic analysis...")
def preprocess_text_for_nlp(text): # ENSURE THIS IS THE VERSION THAT PASSES UNIT TESTS
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r'<script[^>]*>.*?</script>', '', text, flags=re.DOTALL | re.IGNORECASE)
    text = re.sub(r'<style[^>]*>.*?</style>', '', text, flags=re.DOTALL | re.IGNORECASE)
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^\w\s\.\-]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def create_text_from_fields(row, field_weights):
    texts_to_combine = []
    for field, weight in field_weights.items():
        if field not in row: continue
        content_val = row[field]
        actual_content_for_processing = []
        is_list_like_input = False
        if isinstance(content_val, (pd.Series, np.ndarray)):
            actual_content_for_processing = [item for item in content_val if pd.notna(item)]
            is_list_like_input = True
        elif isinstance(content_val, list):
            actual_content_for_processing = [item for item in content_val if pd.notna(item)]
            is_list_like_input = True
        elif pd.notna(content_val):
            actual_content_for_processing = [str(content_val)]
        else: continue
        if weight > 0 and actual_content_for_processing:
            content_str = " ".join(map(str, actual_content_for_processing)) if is_list_like_input else str(actual_content_for_processing[0])
            preprocessed_content_str = preprocess_text_for_nlp(content_str)
            if preprocessed_content_str:
                texts_to_combine.extend([preprocessed_content_str] * int(np.ceil(weight)))
    return " ".join(texts_to_combine)

TEXT_STRATEGIES = {
    'TitleAbstract': {'Fetched_Title': 1, 'Fetched_Abstract': 1},
    'TitleAbstractMeSH': {'Fetched_Title': 1, 'Fetched_Abstract': 1, 'Fetched_MeSH_Terms': 1},
    'TitleEmphasis': {'Fetched_Title': 3, 'Fetched_Abstract': 1},
    'AbstractOnly': {'Fetched_Abstract': 1},
    'FullInputText': {'Title': 1, 'Study Aim/Focus': 1, 'P: Population Details (Age, N, Condition/Surgery Type, Key Characteristics)': 0.5, 'I: Intervention(s) Details (ERAS Protocol Elements, Anesthetic Techniques, eHealth/App Use, Specific Drugs/Dosages, Other Interventions)': 0.5, 'O: Key Outcomes Measured & Findings (LOS, Pain Scores, Opioid Use, Complications, Functional Recovery, Adherence, Patient Satisfaction, Cost, Specific Findings Relevant to Proposal)': 0.5}
}
for strategy_name, field_cfg in TEXT_STRATEGIES.items():
    col_name = f"Text_{strategy_name}"
    df_main[col_name] = df_main.apply(lambda row: create_text_from_fields(row, field_cfg), axis=1)
    print(f"  Created text column: {col_name}")
    print(df_main[[col_name]].head(1).to_string()) # CHANGED

KEYWORD_SOURCE_TEXT_COL = 'Text_TitleAbstract'
if KEYWORD_SOURCE_TEXT_COL not in df_main.columns:
    print(f"Warning: {KEYWORD_SOURCE_TEXT_COL} not found for TF-IDF. Skipping keyword extraction.")
    df_main['TFIDF_Keywords'] = [[] for _ in range(len(df_main))]
else:
    print(f"\nExtracting TF-IDF keywords from '{KEYWORD_SOURCE_TEXT_COL}'...")
    tfidf_vectorizer = TfidfVectorizer(stop_words=STOP_WORDS, max_features=5000, ngram_range=(1,2))
    try:
        valid_texts_for_tfidf = df_main[KEYWORD_SOURCE_TEXT_COL].fillna("")
        if not valid_texts_for_tfidf.str.strip().any(): # Check if all are empty after fillna and strip
             raise ValueError("All documents are empty or only whitespace for TF-IDF.")
        tfidf_matrix = tfidf_vectorizer.fit_transform(valid_texts_for_tfidf)
        feature_names = tfidf_vectorizer.get_feature_names_out()
        def get_top_tfidf_keywords(row_vector, n_top=NUM_KEYWORDS_TFIDF):
            sorted_indices = np.argsort(row_vector.toarray().flatten())[::-1]
            return [feature_names[i] for i in sorted_indices[:n_top] if row_vector[0, i] > 0]
        df_main['TFIDF_Keywords'] = [get_top_tfidf_keywords(tfidf_matrix[i]) for i in range(tfidf_matrix.shape[0])]
        print("  TF-IDF keywords extracted.")
        print(df_main[['PMID', 'TFIDF_Keywords']].head(2).to_string()) # CHANGED
    except ValueError as e:
        print(f"  Error during TF-IDF: {e}. Skipping keyword extraction.")
        df_main['TFIDF_Keywords'] = [[] for _ in range(len(df_main))]

PROCESSED_PROPOSAL_QUERY_TEXT = preprocess_text_for_nlp(PROPOSAL_QUERY_TEXT)
if KEYWORD_SOURCE_TEXT_COL in df_main.columns and 'tfidf_vectorizer' in locals() and PROCESSED_PROPOSAL_QUERY_TEXT.strip():
    try:
        proposal_tfidf_vector = tfidf_vectorizer.transform([PROCESSED_PROPOSAL_QUERY_TEXT])
        PROPOSAL_TFIDF_KEYWORDS = get_top_tfidf_keywords(proposal_tfidf_vector[0])
        print(f"\nTF-IDF Keywords for PROPOSAL_QUERY_TEXT: {PROPOSAL_TFIDF_KEYWORDS}")
    except ValueError: PROPOSAL_TFIDF_KEYWORDS = []; print("Could not extract TF-IDF for proposal (e.g., too short).")
else: PROPOSAL_TFIDF_KEYWORDS = []; print("TF-IDF vectorizer not available or proposal text empty for keyword extraction.")
print("Cell 5: Text Preparation Complete.")

# =============================================================================
# CELL 6: Semantic Similarity Metrics Calculation
# =============================================================================
print("\nCalculating Semantic Similarity Scores...")
try:
    semantic_model = SentenceTransformer(SEMANTIC_MODEL_NAME)
    print(f"  Loaded SentenceTransformer model: {SEMANTIC_MODEL_NAME}")
    proposal_embedding_main = semantic_model.encode(PROCESSED_PROPOSAL_QUERY_TEXT, convert_to_tensor=True)
    for strategy_name in TEXT_STRATEGIES.keys():
        text_col = f"Text_{strategy_name}"; score_col = f"SemSim_{strategy_name}"
        print(f"  Calculating semantic similarity for strategy: {strategy_name} (using '{text_col}')...")
        valid_texts_series = df_main[text_col][df_main[text_col].str.strip() != ""]
        if valid_texts_series.empty:
            print(f"    No valid text for {strategy_name}. Skipping."); df_main[score_col] = np.nan; continue
        doc_embeddings = semantic_model.encode(valid_texts_series.tolist(), convert_to_tensor=True, show_progress_bar=False)
        cosine_scores = util.cos_sim(proposal_embedding_main, doc_embeddings)
        scores_to_assign = np.full(len(df_main), np.nan)
        scores_to_assign[valid_texts_series.index] = cosine_scores[0].cpu().numpy()
        df_main[score_col] = scores_to_assign
        print(f"    {score_col} calculated.")

    if PROPOSAL_TFIDF_KEYWORDS and 'TFIDF_Keywords' in df_main.columns:
        print("  Calculating semantic similarity based on TF-IDF Keywords...")
        if PROPOSAL_TFIDF_KEYWORDS:
            proposal_keyword_embeddings = semantic_model.encode(PROPOSAL_TFIDF_KEYWORDS, convert_to_tensor=True)
            avg_proposal_keyword_embedding = torch.mean(proposal_keyword_embeddings, dim=0, keepdim=True)
        else: avg_proposal_keyword_embedding = None; print("    Proposal has no TF-IDF keywords. Skipping.")
        def calculate_keyword_similarity(doc_keywords_list):
            if not doc_keywords_list or avg_proposal_keyword_embedding is None: return np.nan
            doc_keyword_embeddings = semantic_model.encode(doc_keywords_list, convert_to_tensor=True)
            if doc_keyword_embeddings.nelement() == 0: return np.nan
            avg_doc_keyword_embedding = torch.mean(doc_keyword_embeddings, dim=0, keepdim=True)
            return util.cos_sim(avg_proposal_keyword_embedding, avg_doc_keyword_embedding)[0,0].item()
        df_main['SemSim_TFIDF_Keywords'] = df_main['TFIDF_Keywords'].apply(calculate_keyword_similarity)
        print("    SemSim_TFIDF_Keywords calculated.")
    else: print("  Skipping TF-IDF Keyword-based semantic similarity."); df_main['SemSim_TFIDF_Keywords'] = np.nan
except Exception as e:
    print(f"ERROR during Semantic Similarity calculation: {e}")
    for strategy_name in TEXT_STRATEGIES.keys(): df_main[f"SemSim_{strategy_name}"] = np.nan
    df_main['SemSim_TFIDF_Keywords'] = np.nan

print(df_main[[col for col in df_main.columns if "SemSim_" in col]].head().to_string()) # CHANGED
print("Cell 6: Semantic Similarity Metrics Complete.")

# =============================================================================
# CELL 7: Naive Bayes Classifier
# =============================================================================
print("\nAttempting Naive Bayes Classification for Relevance...")
NB_TEXT_INPUT_COL = 'Text_TitleAbstractMeSH'; NB_TARGET_COL = 'Relevance Score (1-10)'
if NB_TEXT_INPUT_COL not in df_main.columns or NB_TARGET_COL not in df_main.columns:
    print(f"  Input or Target column for Naive Bayes not found. Skipping."); df_main['NB_High_Relevance_Prob'] = np.nan
else:
    df_main['Relevance_Bin'] = pd.cut(df_main[NB_TARGET_COL], bins=RELEVANCE_BINS, labels=RELEVANCE_LABELS, right=True, include_lowest=True)
    value_counts = df_main['Relevance_Bin'].value_counts()
    valid_classes = value_counts[value_counts >= MIN_SAMPLES_PER_CLASS_NB]
    if len(valid_classes) < 2:
        print(f"  Skipping Naive Bayes: Insufficient class diversity. Found: {value_counts.to_dict()}"); df_main['NB_High_Relevance_Prob'] = np.nan
    else:
        print(f"  Sufficient class diversity for Naive Bayes. Bins: {value_counts.to_dict()}")
        df_train_nb = df_main.dropna(subset=[NB_TEXT_INPUT_COL, 'Relevance_Bin'])
        if len(df_train_nb) < (MIN_SAMPLES_PER_CLASS_NB * 2):
            print(f"  Too few samples ({len(df_train_nb)}) after NaNs for Naive Bayes. Skipping."); df_main['NB_High_Relevance_Prob'] = np.nan
        else:
            vectorizer_nb = TfidfVectorizer(stop_words=STOP_WORDS, max_features=5000)
            X_nb = vectorizer_nb.fit_transform(df_train_nb[NB_TEXT_INPUT_COL])
            y_nb = df_train_nb['Relevance_Bin']
            try:
                nb_model = MultinomialNB(); nb_model.fit(X_nb, y_nb)
                valid_text_for_nb_indices = df_main[df_main[NB_TEXT_INPUT_COL].str.strip() != ""].index
                if not valid_text_for_nb_indices.empty:
                    X_pred_nb = vectorizer_nb.transform(df_main.loc[valid_text_for_nb_indices, NB_TEXT_INPUT_COL])
                    predicted_probs_for_valid = nb_model.predict_proba(X_pred_nb)
                    temp_probs_df = pd.DataFrame(predicted_probs_for_valid, index=valid_text_for_nb_indices, columns=nb_model.classes_)
                    df_main['NB_High_Relevance_Prob'] = temp_probs_df.get('High', pd.Series(0.0, index=temp_probs_df.index)) if 'High' in temp_probs_df else 0.0
                    if 'High' not in nb_model.classes_: print("    Warning: 'High' class not in NB model.")
                else: df_main['NB_High_Relevance_Prob'] = np.nan
                print("  Naive Bayes probabilities calculated.")
            except ValueError as ve: print(f"    Skipping NB due to ValueError: {ve}"); df_main['NB_High_Relevance_Prob'] = np.nan
print("Cell 7: Naive Bayes Classification Complete.")

# =============================================================================
# CELL 8: Structural Metrics Calculation
# =============================================================================
print("\nCalculating Structural Metrics (Co-citation, Bibliographic Coupling)...")
df_core_papers_struct = df_main[df_main[NB_TARGET_COL] >= CORE_RELEVANCE_THRESHOLD_FOR_STRUCTURAL]
core_set_pmids_struct = df_core_papers_struct['PMID'].tolist()
if core_set_pmids_struct: print(f"  Core set for structural metrics: {len(core_set_pmids_struct)} PMIDs.")
else: print("  Warning: No core set for structural metrics. Scores will be 0 or NaN.")
pmid_to_date_map_struct = df_main.dropna(subset=['Fetched_PubDate', 'PMID']).set_index('PMID')['Fetched_PubDate'].to_dict()

def calculate_modified_cocitation(target_pmid, target_cited_by_pmids, core_pmids_dict, pmid_date_map):
    if not core_pmids_dict or target_pmid in core_pmids_dict or not target_cited_by_pmids: return 0.0
    target_cited_by_set = set(target_cited_by_pmids); total_score = 0.0
    for core_pmid, core_cited_by_pmids_list in core_pmids_dict.items():
        if target_pmid == core_pmid: continue
        common_citers = target_cited_by_set.intersection(set(core_cited_by_pmids_list)); pair_score = 0.0
        for citer_pmid in common_citers:
            dt_c, dt_t, dt_core = pmid_date_map.get(citer_pmid), pmid_date_map.get(target_pmid), pmid_date_map.get(core_pmid)
            if dt_c and dt_t and dt_core:
                try:
                    age_t = (dt_c - dt_t).days / 365.25; age_c = (dt_c - dt_core).days / 365.25
                    if age_t >= 0 and age_c >= 0: pair_score += 1.0 / ((age_t + 1) * (age_c + 1))
                except TypeError: pass
        total_score += pair_score
    return total_score
core_pmids_cocit_dict = {row['PMID']: row['Fetched_Cited_By_PMIDs'] for _, row in df_core_papers_struct.iterrows()} if core_set_pmids_struct else {}
df_main['Struct_Mod_CoCite'] = df_main.apply(lambda r: calculate_modified_cocitation(r['PMID'], r['Fetched_Cited_By_PMIDs'], core_pmids_cocit_dict, pmid_to_date_map_struct), axis=1)
print("  Modified Co-citation scores calculated.")

def calculate_bibliographic_coupling(target_pmid, target_refs_pmids, core_pmids_refs_dict):
    if not core_pmids_refs_dict or target_pmid in core_pmids_refs_dict or not target_refs_pmids: return 0
    target_refs_set = set(target_refs_pmids); total_score = 0
    for core_pmid, core_refs_pmids_list in core_pmids_refs_dict.items():
        if target_pmid == core_pmid: continue
        total_score += len(target_refs_set.intersection(set(core_refs_pmids_list)))
    return total_score
core_pmids_bibcoup_dict = {row['PMID']: row['Fetched_References_PMIDs_Combined'] for _, row in df_core_papers_struct.iterrows()} if core_set_pmids_struct else {}
df_main['Struct_Bib_Couple'] = df_main.apply(lambda r: calculate_bibliographic_coupling(r['PMID'], r['Fetched_References_PMIDs_Combined'], core_pmids_bibcoup_dict), axis=1)
print("  Bibliographic Coupling scores calculated.")
print(df_main[['PMID', 'Struct_Mod_CoCite', 'Struct_Bib_Couple']].head().to_string()) # CHANGED
print("Cell 8: Structural Metrics Calculation Complete.")

# =============================================================================
# CELL 9: Benchmarking & Correlation Analysis
# =============================================================================
print("\n--- Benchmarking and Correlation Analysis ---")
ground_truth_col = 'Relevance Score (1-10)'
metric_cols_for_corr = [col for col in df_main.columns if "SemSim_" in col or "NB_" in col or "Struct_" in col]
if not metric_cols_for_corr or ground_truth_col not in df_main.columns:
    print("No metric columns or ground truth column found for correlation. Skipping.")
else:
    df_for_corr = df_main[[ground_truth_col] + metric_cols_for_corr].copy()
    for col in df_for_corr.columns: df_for_corr[col] = pd.to_numeric(df_for_corr[col], errors='coerce')
    correlations = df_for_corr.corr(method='spearman')
    print(f"\nSpearman Correlation with '{ground_truth_col}':")
    ground_truth_corrs = correlations[[ground_truth_col]].drop(ground_truth_col, errors='ignore').sort_values(by=ground_truth_col, ascending=False) # Added errors='ignore'
    print(ground_truth_corrs.to_string()) # CHANGED

    plt.figure(figsize=(max(10, len(metric_cols_for_corr)*0.6) , max(8, len(metric_cols_for_corr)*0.5)))
    sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1, annot_kws={"size": 8})
    plt.title(f'Spearman Correlation Heatmap (vs. {ground_truth_col})', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(rotation=0, fontsize=10)
    plt.tight_layout()
    plt.show()
print("Cell 9: Benchmarking and Correlation Analysis Complete.")

# =============================================================================
# CELL 10: Plotting Overhauled - Metrics vs. Ground Truth & Distributions
# =============================================================================
print("\n--- Plotting Overhauled: Metrics vs. Ground Truth & Distributions ---")
if not metric_cols_for_corr or ground_truth_col not in df_main.columns:
    print("No metric columns or ground truth column for plotting. Skipping.")
else:
    num_metrics_plot = len(metric_cols_for_corr)
    if num_metrics_plot > 0:
        # --- Figure 1: Metrics vs. Ground Truth (Scatter + Regression) ---
        print("\nFigure 1: Scatter plots with Regression Lines (Metric vs. Ground Truth)")
        cols_subplot = 2
        rows_subplot = (num_metrics_plot + cols_subplot - 1) // cols_subplot
        fig1, axes1 = plt.subplots(nrows=rows_subplot, ncols=cols_subplot, figsize=(16, 6 * rows_subplot), squeeze=False) # squeeze=False ensures axes is 2D
        axes1 = axes1.flatten()

        for i, metric in enumerate(metric_cols_for_corr):
            ax = axes1[i]
            if metric in df_main.columns and df_main[metric].notna().any():
                # Drop NaN from both metric and ground truth for this specific plot pair
                plot_data = df_main[[metric, ground_truth_col]].dropna()
                if not plot_data.empty:
                    sns.regplot(data=plot_data, x=metric, y=ground_truth_col, ax=ax,
                                scatter_kws={'alpha':0.6, 's': 50},
                                line_kws={'color':'red', 'linestyle':'--'})
                    
                    # Calculate Spearman correlation for annotation
                    rho, p_value = scipy.stats.spearmanr(plot_data[metric], plot_data[ground_truth_col], nan_policy='omit')
                    annotation = f"ρ = {rho:.2f}\np = {p_value:.2g}" # Using .2g for p-value for scientific notation if small
                    ax.text(0.05, 0.95, annotation, transform=ax.transAxes, fontsize=9,
                            verticalalignment='top', bbox=dict(boxstyle='round,pad=0.3', fc='wheat', alpha=0.5))
                    
                    ax.set_title(f'{metric} vs. {ground_truth_col}', fontsize=12)
                    ax.set_xlabel(metric, fontsize=10)
                    ax.set_ylabel(ground_truth_col, fontsize=10)
                    ax.grid(True, linestyle=':', alpha=0.7)
                else:
                    ax.set_title(f'{metric}\n(No non-NaN data pairs)')
                    ax.axis('off')
            else:
                ax.set_title(f'{metric}\n(No data or all NaN)')
                ax.axis('off')
        
        for j in range(i + 1, len(axes1)): fig1.delaxes(axes1[j]) # Hide unused subplots
        fig1.suptitle("Metrics vs. Ground Truth Relevance Score", fontsize=16, y=1.02)
        fig1.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to make space for suptitle
        plt.show()

        # --- Figure 2: Distributions of Calculated Metrics ---
        print("\nFigure 2: Distributions of Calculated Metrics (Histograms & KDEs)")
        fig2, axes2 = plt.subplots(nrows=rows_subplot, ncols=cols_subplot, figsize=(16, 6 * rows_subplot), squeeze=False)
        axes2 = axes2.flatten()

        for i, metric in enumerate(metric_cols_for_corr):
            ax = axes2[i]
            if metric in df_main.columns and df_main[metric].notna().any():
                metric_data = df_main[metric].dropna()
                if not metric_data.empty:
                    sns.histplot(metric_data, kde=True, ax=ax, bins=15, stat="density")
                    ax.set_title(f'Distribution of {metric}', fontsize=12)
                    ax.set_xlabel(metric, fontsize=10)
                    ax.set_ylabel('Density', fontsize=10)
                    ax.grid(True, linestyle=':', alpha=0.7)
                else:
                    ax.set_title(f'Distribution of {metric}\n(No non-NaN data)')
                    ax.axis('off')
            else:
                ax.set_title(f'Distribution of {metric}\n(No data or all NaN)')
                ax.axis('off')

        for j in range(i + 1, len(axes2)): fig2.delaxes(axes2[j]) # Hide unused subplots
        fig2.suptitle("Distributions of Calculated Metrics", fontsize=16, y=1.02)
        fig2.tight_layout(rect=[0, 0, 1, 0.98])
        plt.show()
    else:
        print("No valid metrics available for plotting.")
print("Cell 10: Plotting Overhauled Complete.")

# =============================================================================
# CELL 11: Display Final DataFrame Sample
# =============================================================================
print("\n--- Final DataFrame Sample with All Calculated Scores ---")
display_cols_final = ['PMID', 'Title', ground_truth_col] + \
                     ['Fetched_Title', 'Fetched_DOI', 'Fetched_PubDate'] + \
                     metric_cols_for_corr + \
                     ['Fetched_References_PMIDs_Combined', 'Fetched_Cited_By_PMIDs']
final_cols_to_show = [col for col in display_cols_final if col in df_main.columns]

# To control width and make it more readable for very wide DataFrames
# This will print it in a way that's easier to copy, but might be long.
with pd.option_context('display.max_rows', 10, 'display.max_columns', None, 'display.width', 2000, 'display.max_colwidth', 70): # CHANGED
    print(df_main[final_cols_to_show].head().to_string())

# Optional: Save to new Excel
# output_excel_path = "screening_ERAS_augmented_and_scored_vNext.xlsx"
# df_main.to_excel(output_excel_path, index=False)
# print(f"\nFull augmented DataFrame saved to: {output_excel_path}")

print("\n--- Analysis Script Fully Complete ---")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
import time
import re
from urllib.parse import urljoin, urlparse
from urllib3.util.retry import Retry
import xml.etree.ElementTree as ET
from requests.adapters import HTTPAdapter
from concurrent.futures import ThreadPoolExecutor, as_completed
import logging

# --- Configuration ---
EXCEL_FILE_PATH = r"C:\Users\Galaxy\Downloads\screening_ERAS.xlsx" # Use 'r' for raw string or double backslashes
OUTPUT_PDF_DIR = "downloaded_pdfs_with_doi_parallel"

# Sci-Hub Configuration
# Updated list based on recent search results & common knowledge
SCI_HUB_DOMAINS = [
    "https://sci-hub.se",
    "https://sci-hub.ru",
    "https://sci-hub.st", # Was giving 403s, but often listed. Keep for now.
    "https://sci-hub.wf",
    "https://sci-hub.ee",
    "https://sci-hub.ren",
    "https://sci-hub.cat",
    "https://sci-hub.yt", # Found in some lists, worth trying
    "https://sci-hub.et",
    "https://sci-hub.shop",
    # Add more known working domains here if needed
]
REQUEST_DELAY_SCIHUB = 3 # Seconds to wait between trying different Sci-Hub domains for the *same* article
# Overall politeness is now managed by number of workers and this internal delay.
MAX_WORKERS_SCIHUB = 5 # Number of parallel download threads for Sci-Hub

# NCBI E-utilities Configuration
NCBI_API_KEY = "YOUR_API_KEY_HERE"  # Get from: https://www.ncbi.nlm.nih.gov/account/settings/
CROSSREF_MAILTO = "your_email@example.com" # Your actual email address

EFETCH_BATCH_SIZE = 100
REQUEST_DELAY_NCBI = 0.35 # Base delay for NCBI requests

# --- Logging ---
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# --- NCBI Session & Helper ---
session_ncbi = requests.Session()

# Configure retries for the NCBI session
retries_ncbi = Retry(
    total=3,
    backoff_factor=0.5,
    status_forcelist=[500, 502, 503, 504],
    allowed_methods=frozenset(['GET', 'POST']) # Be explicit about methods if needed
)
adapter_ncbi = HTTPAdapter(
    pool_connections=10,
    pool_maxsize=10,
    max_retries=retries_ncbi # Pass the Retry object here
)
session_ncbi.mount("https://", adapter_ncbi)
session_ncbi.mount("http://", adapter_ncbi)
ncbi_user_agent_base = 'SciHubPDFDownloader/2.1 (Python Requests; mailto:{})'
session_ncbi.headers.update({'User-Agent': ncbi_user_agent_base.format(CROSSREF_MAILTO if CROSSREF_MAILTO != "your_email@example.com" else "anonymous_user")})

def _get_ncbi_params(extra=None):
    params = {"tool": "pdf_downloader_eras_v2", "email": CROSSREF_MAILTO}
    if NCBI_API_KEY and NCBI_API_KEY != "YOUR_API_KEY_HERE":
        params["api_key"] = NCBI_API_KEY
    if CROSSREF_MAILTO == "your_email@example.com" and "email" in params:
         logger.warning("`CROSSREF_MAILTO` is placeholder. Provide actual email for NCBI politeness.")
         params.pop("email") # Remove placeholder if it's still default
    if extra:
        params.update(extra)
    return params

def fetch_dois_for_pmids_batched(pmids_list, batch_size=EFETCH_BATCH_SIZE):
    t_start_fetch_doi = time.time()
    pmid_to_doi_map = {pmid: None for pmid in pmids_list}
    valid_pmids_to_query = [pmid for pmid in pmids_list if pmid]

    if not valid_pmids_to_query:
        logger.info("No valid PMIDs provided to fetch DOIs.")
        return pmid_to_doi_map

    logger.info(f"Fetching DOIs for {len(valid_pmids_to_query)} PMIDs from NCBI (batch size: {batch_size})...")
    
    delay_ncbi = REQUEST_DELAY_NCBI
    if not (NCBI_API_KEY and NCBI_API_KEY != "YOUR_API_KEY_HERE"):
        logger.info("NCBI API Key not provided or is placeholder. Using default delay for NCBI requests.")
        # NCBI rate limits are 3/sec without API key, 10/sec with.
        # Delay is already 0.35s, which is ~3 req/s.
    
    # Update session User-Agent with the actual email for NCBI
    session_ncbi.headers.update({'User-Agent': ncbi_user_agent_base.format(CROSSREF_MAILTO if CROSSREF_MAILTO != "your_email@example.com" else "anonymous_user")})


    for i in range(0, len(valid_pmids_to_query), batch_size):
        batch_pmids = valid_pmids_to_query[i:i + batch_size]
        batch_num = i // batch_size + 1
        logger.info(f"  Fetching DOIs: NCBI EFetch batch {batch_num} ({len(batch_pmids)} PMIDs)")
        
        post_data = _get_ncbi_params({"db": "pubmed", "retmode": "xml", "id": ",".join(batch_pmids)}) # rettype=abstract implied by default
        
        try:
            response = session_ncbi.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi", data=post_data, timeout=60)
            response.raise_for_status()

            if not response.content:
                logger.warning(f"   - EFetch returned empty content for NCBI batch {batch_num}.")
                continue

            root = ET.fromstring(response.content)
            for art_node in root.findall(".//PubmedArticle"): # Top level for each article
                pmid_elem = art_node.find(".//MedlineCitation/PMID")
                pmid = pmid_elem.text if pmid_elem is not None else None
                if not pmid: continue

                doi = None
                # Path from provided DAG checker: MedlineCitation/Article/ELocationID[@EIdType='doi'][@ValidYN='Y']
                # Path from general observation: PubmedData/ArticleIdList/ArticleId[@IdType='doi']
                # Let's try both, prioritizing the one that's usually more direct
                doi_elem = art_node.find(f".//ArticleIdList/ArticleId[@IdType='doi']") # Inside PubmedArticle or PubmedBookArticle
                if doi_elem is None: # Fallback for some structures like in the DAG checker example
                     doi_elem = art_node.find(f".//MedlineCitation/Article/ELocationID[@EIdType='doi'][@ValidYN='Y']")

                if doi_elem is not None and doi_elem.text:
                    doi = doi_elem.text.strip()
                
                if pmid in pmid_to_doi_map and doi:
                    pmid_to_doi_map[pmid] = doi
        
        except requests.exceptions.HTTPError as e:
            logger.error(f"   - HTTP Error during EFetch for NCBI batch {batch_num}: {e.response.status_code} {e.response.text[:200]}")
        except requests.exceptions.RequestException as e:
            logger.error(f"   - Request Error during EFetch for NCBI batch {batch_num}: {e}")
        except ET.ParseError as e:
            logger.error(f"   - XML Parse Error during EFetch for NCBI batch {batch_num}: {e}")
        except Exception as e:
            logger.error(f"   - Unexpected Error during EFetch for NCBI batch {batch_num}: {e}", exc_info=True)
        
        time.sleep(delay_ncbi)

    t_end_fetch_doi = time.time()
    found_doi_count = sum(1 for doi_val in pmid_to_doi_map.values() if doi_val)
    logger.info(f"Finished fetching DOIs. Found for {found_doi_count}/{len(valid_pmids_to_query)} PMIDs. (Took {t_end_fetch_doi - t_start_fetch_doi:.2f}s)")
    return pmid_to_doi_map

# --- Sci-Hub Session & Helpers ---
session_scihub = requests.Session()
# Using a more common browser User-Agent
scihub_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'Accept-Language': 'en-US,en;q=0.9',
    'DNT': '1', # Do Not Track
    'Upgrade-Insecure-Requests': '1'
}
session_scihub.headers.update(scihub_headers)

# Configure retries for the Sci-Hub session
retries_scihub = Retry(
    total=2, # As you had retry_total=2
    backoff_factor=0.5,
    status_forcelist=[500, 502, 503, 504], # Common server errors
    # For Sci-Hub, we mostly use GET
    allowed_methods=frozenset(['GET'])
)
adapter_scihub = HTTPAdapter(
    pool_connections=MAX_WORKERS_SCIHUB + 5,
    pool_maxsize=MAX_WORKERS_SCIHUB + 5,
    max_retries=retries_scihub # Pass the Retry object here
)
session_scihub.mount("https://", adapter_scihub)
session_scihub.mount("http://", adapter_scihub)


def sanitize_filename(filename):
    filename = re.sub(r'[\\/*?:"<>|]', "", filename)
    filename = filename.replace(" ", "_")
    return filename[:200]

def find_pdf_url_on_scihub_page(html_content, scihub_page_url):
    soup = BeautifulSoup(html_content, 'html.parser')
    pdf_url = None
    scihub_domain_base = f"{urlparse(scihub_page_url).scheme}://{urlparse(scihub_page_url).netloc}"

    # Most common: iframe with id 'pdf' or 'article'
    for iframe_id in ['pdf', 'article']:
        iframe = soup.find('iframe', id=iframe_id)
        if iframe and iframe.get('src'):
            pdf_url = iframe['src']
            break
    if not pdf_url: # Try embed tag
        embed = soup.find('embed', type='application/pdf')
        if embed and embed.get('src'):
            pdf_url = embed['src']

    # Try direct links or buttons if iframe/embed not found
    if not pdf_url:
        # Look for download buttons with onclick handlers
        # Example: <button onclick="location.href='...'"></button>
        # Or <div onclick="location.href='...'" id="download">
        clickable_elements = soup.find_all(['button', 'a', 'div'], onclick=True)
        for elem in clickable_elements:
            onclick_attr = elem['onclick']
            match = re.search(r"location\.href\s*=\s*['\"]([^'\"]+\.pdf[^'\"]*)['\"]", onclick_attr) # PDF link
            if not match:
                 match = re.search(r"location\.href\s*=\s*['\"]([^'\"]+)['\"]", onclick_attr) # Any link
            if match:
                potential_url = match.group(1)
                # Sci-Hub sometimes has links like /downloads/.... or //another.domain/....
                if '.pdf' in potential_url or 'download=true' in potential_url or 'sci-hub.ee/downloads' in potential_url : # Added another common pattern
                    pdf_url = potential_url
                    break
    
    if not pdf_url:
        # Look for <a> tags with href pointing to PDF or common Sci-Hub paths
        for a_tag in soup.find_all('a', href=True):
            href = a_tag['href']
            # Check for .pdf, common Sci-Hub paths, or parameters indicating download
            if ('.pdf' in href.lower() or
                'plugin=1' in href or
                '/downloads/' in href or
                'sci-hub.wf/downloads/' in href or # common pattern for some mirrors
                (a_tag.get('id') == 'download' and href != '#') or
                (a_tag.text.lower().strip() == 'save' and href != '#') or
                (a_tag.text.lower().strip() == 'download' and href != '#')):
                pdf_url = href
                break

    if pdf_url:
        if pdf_url.startswith("//"):
            pdf_url = urlparse(scihub_page_url).scheme + ":" + pdf_url # Use scheme from original page
        elif not pdf_url.startswith("http"):
            pdf_url = urljoin(scihub_domain_base, pdf_url)
        return pdf_url
        
    logger.debug(f"    [DEBUG] Could not find PDF link using common patterns on Sci-Hub page: {scihub_page_url}")
    return None

def download_pdf_from_scihub_attempt(identifier, identifier_type, pmid_for_filename, first_author, year, output_dir):
    base_filename = f"{year}_{first_author}_{pmid_for_filename}"
    pdf_filename = sanitize_filename(base_filename) + ".pdf"
    pdf_filepath = os.path.join(output_dir, pdf_filename)

    if os.path.exists(pdf_filepath):
        logger.info(f"  PDF already exists: {pdf_filepath} (PMID: {pmid_for_filename})")
        return True # Success

    logger.info(f"  Attempting Sci-Hub for {identifier_type}: {identifier} (PMID: {pmid_for_filename})")
    
    for scihub_domain in SCI_HUB_DOMAINS:
        search_url = f"{scihub_domain.strip('/')}/{identifier}" # Ensure no double slashes
        logger.info(f"    Trying Sci-Hub domain: {search_url}")

        try:
            time.sleep(REQUEST_DELAY_SCIHUB) # Small delay before each domain attempt
            
            # Initial request to Sci-Hub to get the page that might contain the PDF link or redirect
            response = session_scihub.get(search_url, timeout=30) # Increased timeout
            response.raise_for_status()
            content_type = response.headers.get('Content-Type', '').lower()

            # Case 1: Sci-Hub directly returns the PDF (less common now)
            if 'application/pdf' in content_type:
                logger.info(f"      Direct PDF found from {search_url}!")
                with open(pdf_filepath, 'wb') as f:
                    f.write(response.content)
                logger.info(f"      SUCCESS: Downloaded {pdf_filename} (via {identifier_type} from {scihub_domain})")
                return True
            
            # Case 2: Sci-Hub returns an HTML page, parse it
            elif 'text/html' in content_type:
                logger.info(f"      HTML page received from {search_url}. Parsing for PDF link...")
                # Pass the actual URL from which HTML was fetched for correct base URL joining
                direct_pdf_url = find_pdf_url_on_scihub_page(response.content, response.url) 
                
                if direct_pdf_url:
                    logger.info(f"      Found potential PDF URL: {direct_pdf_url}")
                    # Another delay before downloading the actual PDF is good practice, but might be handled by overall worker flow
                    # time.sleep(REQUEST_DELAY_SCIHUB / 2) # Optional shorter delay
                    
                    pdf_response = session_scihub.get(direct_pdf_url, timeout=60, stream=True)
                    pdf_response.raise_for_status()
                    
                    pdf_content_type = pdf_response.headers.get('Content-Type', '').lower()
                    if 'application/pdf' in pdf_content_type:
                        with open(pdf_filepath, 'wb') as f:
                            for chunk in pdf_response.iter_content(chunk_size=8192):
                                f.write(chunk)
                        logger.info(f"      SUCCESS: Downloaded {pdf_filename} (via {identifier_type} from {direct_pdf_url})")
                        return True
                    else:
                        logger.warning(f"      Link found ({direct_pdf_url}) did not lead to a PDF. Content-Type: {pdf_content_type}")
                        # Optionally save the content for inspection:
                        # with open(pdf_filepath + f".unexpected_type_from_{urlparse(direct_pdf_url).hostname}.html", 'wb') as f:
                        #     f.write(pdf_response.content)
                else:
                    logger.warning(f"      Could not find PDF link on HTML page from {search_url}.")
            else:
                logger.warning(f"      Unexpected Content-Type '{content_type}' from {search_url}. Skipping this domain for this item.")

        except requests.exceptions.HTTPError as e:
            if e.response.status_code == 404:
                logger.warning(f"      Article {identifier} not found on {scihub_domain} (404 Error).")
            elif e.response.status_code == 403:
                logger.warning(f"      Access Forbidden (403 Error) for {search_url}. This domain might be blocking.")
            else:
                logger.error(f"      HTTP Error for {search_url}: {e.response.status_code} {e.response.reason}")
        except requests.exceptions.Timeout:
            logger.error(f"      Timeout for {search_url}.")
        except requests.exceptions.RequestException as e:
            logger.error(f"      Request failed for {search_url}: {e}")
        except Exception as e:
            logger.error(f"      An unexpected error occurred with {search_url}: {e}", exc_info=True)
            
    return False # Failed for this identifier after trying all domains

def process_article_row(args):
    """
    Worker function to process a single article row (get DOI if needed, download PDF).
    `args` is a tuple: (index, row_data, pmid_to_doi_map, output_dir)
    """
    index, row, pmid_to_doi_map, output_dir = args
    pmid_str = str(row.get('PMID', '')).strip().split('.')[0]
    
    if not pmid_str:
        logger.info(f"Skipping Excel row {index + 2}: No PMID.")
        return pmid_str, False # pmid, success_status

    first_author = str(row.get('First Author', 'UnknownAuthor')).strip()
    year = str(row.get('Year', 'UnknownYear')).strip()
    
    logger.info(f"Processing article (Excel row {index+2}): PMID {pmid_str} ({first_author}, {year})")

    article_doi = pmid_to_doi_map.get(pmid_str) # DOI should be pre-fetched
    download_successful = False

    if article_doi:
        logger.info(f"  Using pre-fetched DOI: {article_doi} for PMID {pmid_str}.")
        if download_pdf_from_scihub_attempt(article_doi, "DOI", pmid_str, first_author, year, output_dir):
            download_successful = True
        else:
            logger.warning(f"  DOI ({article_doi}) download failed for PMID {pmid_str}. Attempting with PMID as fallback.")
    else:
        logger.info(f"  No DOI found/pre-fetched for PMID {pmid_str}. Attempting with PMID directly.")

    if not download_successful:
        if download_pdf_from_scihub_attempt(pmid_str, "PMID", pmid_str, first_author, year, output_dir):
            download_successful = True
        else:
            logger.error(f"  PMID ({pmid_str}) download also failed (after DOI attempt if applicable).")
    
    if not download_successful:
        logger.error(f"FAILED to download PDF for PMID {pmid_str} (Excel row {index+2}) using any method.")

    return pmid_str, download_successful


# --- Main Script ---
def main():
    logger.info(f"Starting PDF download process...")
    logger.info(f"Reading Excel file: {EXCEL_FILE_PATH}")

    if NCBI_API_KEY == "YOUR_API_KEY_HERE":
        logger.warning("\n*** NCBI_API_KEY is placeholder. NCBI requests will be slower and more rate-limited. ***")
        logger.warning("*** Get key: https://www.ncbi.nlm.nih.gov/account/settings/ and update script. ***\n")
    if CROSSREF_MAILTO == "your_email@example.com":
        logger.warning("\n*** CROSSREF_MAILTO (for NCBI politeness) is placeholder. Please update with your email. ***\n")

    try:
        df = pd.read_excel(EXCEL_FILE_PATH)
    except FileNotFoundError:
        logger.error(f"Excel file not found: {EXCEL_FILE_PATH}.")
        return
    except Exception as e:
        logger.error(f"Could not read Excel file: {e}", exc_info=True)
        return

    if 'PMID' not in df.columns:
        logger.error("'PMID' column not found in Excel. Available: {df.columns.tolist()}")
        return

    if not os.path.exists(OUTPUT_PDF_DIR):
        os.makedirs(OUTPUT_PDF_DIR)
        logger.info(f"Created output directory: {OUTPUT_PDF_DIR}")

    all_pmids_from_excel = [str(pmid).split('.')[0] for pmid in df['PMID'].dropna().unique() if str(pmid).strip()]
    
    # 1. Fetch all DOIs in batch (sequentially, as this is one large batch process usually)
    pmid_to_doi_map = fetch_dois_for_pmids_batched(all_pmids_from_excel)

    downloaded_count = 0
    failed_count = 0
    
    tasks = []
    for index, row in df.iterrows():
        tasks.append((index, row, pmid_to_doi_map, OUTPUT_PDF_DIR))

    logger.info(f"\n--- Starting Parallel PDF Download Phase (Max Workers: {MAX_WORKERS_SCIHUB}) ---")
    
    with ThreadPoolExecutor(max_workers=MAX_WORKERS_SCIHUB) as executor:
        future_to_pmid = {executor.submit(process_article_row, task_args): task_args[1].get('PMID') for task_args in tasks}
        
        for future in as_completed(future_to_pmid):
            _pmid_excel = future_to_pmid[future] # Original PMID from Excel row for reference
            try:
                _pmid_processed, success = future.result()
                if success:
                    downloaded_count += 1
                else:
                    failed_count += 1
            except Exception as exc:
                logger.error(f"Article with original Excel PMID {_pmid_excel} generated an exception: {exc}", exc_info=True)
                failed_count += 1
                
    logger.info("\n--- Download Summary ---")
    logger.info(f"Total articles in Excel: {len(df)}")
    logger.info(f"Successfully downloaded (or already existed): {downloaded_count} articles.")
    logger.info(f"Failed to download:      {failed_count} articles.")
    logger.info(f"PDFs saved to: {os.path.abspath(OUTPUT_PDF_DIR)}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
import time
import re
from urllib.parse import urljoin, urlparse
from urllib3.util.retry import Retry
import xml.etree.ElementTree as ET
from requests.adapters import HTTPAdapter
from concurrent.futures import ThreadPoolExecutor, as_completed
import logging
import random

# --- Configuration ---
EXCEL_FILE_PATH = r"C:\Users\Galaxy\Downloads\screening_ERAS.xlsx"
OUTPUT_PDF_DIR = "downloaded_pdfs_v3"

# Sci-Hub Configuration
# Prioritize domains that have historically been more stable or are commonly recommended.
# We will try to find a working one from this list.
SCI_HUB_DOMAINS = [
    "https://sci-hub.se",
    "https://sci-hub.ru",
    "https://sci-hub.ren", # Often cited as working
    "https://sci-hub.wf",  # Keep for testing, though problematic in logs
    "https://sci-hub.ee",
    "https://sci-hub.st",  # Keep for testing, though problematic in logs
    # "https://sci-hub.cat", # Can add more if needed
    # "https://sci-hub.yt",
    # "https://sci-hub.shop",
    # "https://sci-hub.et", # Removed due to consistent NameResolutionError
]
REQUEST_DELAY_SCIHUB = 2  # Shorter delay between attempts on *different* domains for the same article
MAX_WORKERS_SCIHUB = 4    # Number of parallel download threads
PREFERRED_DOMAINS = []    # Will be populated by testing
FALLBACK_DOMAINS_COUNT = 2 # How many additional domains to try if preferred ones fail

# NCBI E-utilities Configuration
NCBI_API_KEY = "YOUR_API_KEY_HERE"
CROSSREF_MAILTO = "your_email@example.com"

EFETCH_BATCH_SIZE = 100
REQUEST_DELAY_NCBI = 0.35

# --- Logging ---
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - [%(threadName)s] - %(message)s')
logger = logging.getLogger(__name__)

# --- NCBI Session & Helper ---
session_ncbi = requests.Session()
retries_ncbi = Retry(total=3, backoff_factor=0.5, status_forcelist=[500, 502, 503, 504], allowed_methods=frozenset(['GET', 'POST']))
adapter_ncbi = HTTPAdapter(pool_connections=10, pool_maxsize=10, max_retries=retries_ncbi)
session_ncbi.mount("https://", adapter_ncbi)
session_ncbi.mount("http://", adapter_ncbi)
ncbi_user_agent_base = 'SciHubPDFDownloader/2.2 (Python Requests; mailto:{})'
# User-Agent will be set before first NCBI call in main()

def _get_ncbi_params(extra=None):
    params = {"tool": "pdf_downloader_eras_v3", "email": CROSSREF_MAILTO}
    if NCBI_API_KEY and NCBI_API_KEY != "YOUR_API_KEY_HERE":
        params["api_key"] = NCBI_API_KEY
    if CROSSREF_MAILTO == "your_email@example.com" and "email" in params:
         logger.warning("`CROSSREF_MAILTO` is placeholder. Provide actual email for NCBI politeness.")
         params.pop("email", None)
    if extra: params.update(extra)
    return params

def fetch_dois_for_pmids_batched(pmids_list, batch_size=EFETCH_BATCH_SIZE):
    t_start_fetch_doi = time.time()
    pmid_to_doi_map = {pmid: None for pmid in pmids_list}
    valid_pmids_to_query = [pmid for pmid in pmids_list if pmid]

    if not valid_pmids_to_query:
        logger.info("No valid PMIDs provided to fetch DOIs.")
        return pmid_to_doi_map

    logger.info(f"Fetching DOIs for {len(valid_pmids_to_query)} PMIDs from NCBI (batch size: {batch_size})...")
    delay_val = REQUEST_DELAY_NCBI
    if not (NCBI_API_KEY and NCBI_API_KEY != "YOUR_API_KEY_HERE"):
        logger.info("NCBI API Key not set or placeholder. Using default delay for NCBI requests.")

    for i in range(0, len(valid_pmids_to_query), batch_size):
        batch_pmids = valid_pmids_to_query[i:i + batch_size]
        batch_num = i // batch_size + 1
        logger.info(f"  Fetching DOIs: NCBI EFetch batch {batch_num} ({len(batch_pmids)} PMIDs)")
        post_data = _get_ncbi_params({"db": "pubmed", "retmode": "xml", "id": ",".join(batch_pmids)})
        try:
            response = session_ncbi.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi", data=post_data, timeout=60)
            response.raise_for_status()
            if not response.content:
                logger.warning(f"   - EFetch returned empty content for NCBI batch {batch_num}.")
                continue
            root = ET.fromstring(response.content)
            for art_node in root.findall(".//PubmedArticle"):
                pmid_elem = art_node.find(".//MedlineCitation/PMID")
                pmid = pmid_elem.text if pmid_elem is not None else None
                if not pmid: continue
                doi = None
                doi_elem = art_node.find(f".//ArticleIdList/ArticleId[@IdType='doi']")
                if doi_elem is None:
                     doi_elem = art_node.find(f".//MedlineCitation/Article/ELocationID[@EIdType='doi'][@ValidYN='Y']")
                if doi_elem is not None and doi_elem.text: doi = doi_elem.text.strip()
                if pmid in pmid_to_doi_map and doi: pmid_to_doi_map[pmid] = doi
        except Exception as e:
            logger.error(f"   - Error during EFetch for NCBI batch {batch_num}: {e}", exc_info=False) # Show less detail for batch errors
        time.sleep(delay_val)
    t_end_fetch_doi = time.time()
    found_doi_count = sum(1 for d_val in pmid_to_doi_map.values() if d_val)
    logger.info(f"Finished fetching DOIs. Found for {found_doi_count}/{len(valid_pmids_to_query)} PMIDs. (Took {t_end_fetch_doi - t_start_fetch_doi:.2f}s)")
    return pmid_to_doi_map

# --- Sci-Hub Session & Helpers ---
session_scihub = requests.Session()
scihub_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'Accept-Language': 'en-US,en;q=0.9', 'DNT': '1', 'Upgrade-Insecure-Requests': '1'
}
session_scihub.headers.update(scihub_headers)
retries_scihub = Retry(total=2, backoff_factor=0.3, status_forcelist=[500, 502, 503, 504], allowed_methods=frozenset(['GET']))
adapter_scihub = HTTPAdapter(pool_connections=MAX_WORKERS_SCIHUB + 5, pool_maxsize=MAX_WORKERS_SCIHUB + 5, max_retries=retries_scihub)
session_scihub.mount("https://", adapter_scihub)
session_scihub.mount("http://", adapter_scihub)

def sanitize_filename(filename):
    filename = re.sub(r'[\\/*?:"<>|]', "", filename)
    filename = filename.replace(" ", "_").replace("__", "_")
    return filename[:200]

def find_pdf_url_on_scihub_page(html_content, scihub_page_url):
    soup = BeautifulSoup(html_content, 'html.parser')
    pdf_url = None
    scihub_domain_base = f"{urlparse(scihub_page_url).scheme}://{urlparse(scihub_page_url).netloc}"
    logger.debug(f"Parsing HTML from {scihub_page_url} (base: {scihub_domain_base})")

    # 1. Try iframe with id="pdf" or id="article" (most common for direct embeds)
    for iframe_id_try in ['pdf', 'article']:
        iframe = soup.find('iframe', id=iframe_id_try)
        if iframe and iframe.get('src'):
            logger.debug(f"  Found iframe #{iframe_id_try} with src: {iframe['src']}")
            pdf_url = iframe['src']
            break
    
    # 2. If not found, try embed tag with type application/pdf
    if not pdf_url:
        embed = soup.find('embed', type='application/pdf')
        if embed and embed.get('src'):
            logger.debug(f"  Found embed tag with src: {embed['src']}")
            pdf_url = embed['src']

    # 3. If not found, look for specific download buttons or links by ID or content
    if not pdf_url:
        # Common patterns for download links/buttons
        download_elements = soup.select('button#download, a#download, div#download, button[onclick*="location.href"], a[onclick*="location.href"]')
        # Also elements containing text like "download" or "save"
        for text_pattern in ['download', 'save', '↓', 'get article']: # Added common symbols/phrases
            download_elements.extend(soup.find_all(lambda tag: tag.name in ['a', 'button'] and text_pattern in tag.text.lower()))
        
        for elem in download_elements:
            temp_url = None
            if elem.get('onclick'):
                match = re.search(r"location\.href\s*=\s*['\"]([^'\"]+)['\"]", elem['onclick'])
                if match: temp_url = match.group(1)
            elif elem.get('href') and elem.get('href') != '#':
                temp_url = elem['href']
            
            if temp_url:
                logger.debug(f"  Found potential link/button: '{elem.text.strip()[:30]}' -> {temp_url}")
                # Check if it looks like a PDF or known Sci-Hub download path
                if '.pdf' in temp_url.lower() or 'downloads' in temp_url.lower() or 'tree' in temp_url.lower() or 'sci-hub' in temp_url.lower():
                    pdf_url = temp_url
                    break # Take the first promising one

    if pdf_url:
        # Resolve URL (make absolute)
        if pdf_url.startswith("//"):
            pdf_url = urlparse(scihub_page_url).scheme + ":" + pdf_url
        elif not pdf_url.startswith("http"):
            pdf_url = urljoin(scihub_domain_base, pdf_url)
        logger.debug(f"  Resolved PDF URL: {pdf_url}")
        return pdf_url
    
    logger.warning(f"  [!] Could not find a definitive PDF link on page {scihub_page_url}")
    # Optional: Save HTML for debugging if parsing fails
    # with open(f"debug_page_{time.time()}.html", "w", encoding="utf-8") as f_debug:
    #    f_debug.write(html_content)
    return None


def test_scihub_domain(domain, test_doi="10.1000/182"): # A known non-existent but valid format DOI
    """Tests if a Sci-Hub domain is responsive and likely to serve pages."""
    test_url = f"{domain.strip('/')}/{test_doi}"
    try:
        response = session_scihub.get(test_url, timeout=10) # Shorter timeout for test
        if response.status_code == 200 and 'text/html' in response.headers.get('Content-Type','').lower():
            # Further check if it's a Sci-Hub page, not just any 200
            if "sci-hub" in response.text.lower() or 'pdf' in response.text.lower():
                 logger.info(f"  [Sci-Hub Test] Domain {domain} seems responsive and serves Sci-Hub like HTML.")
                 return True
            else:
                 logger.warning(f"  [Sci-Hub Test] Domain {domain} responsive but content not typical Sci-Hub HTML.")
                 return False
        elif response.status_code == 404 and "article not found" in response.text.lower(): # This is also a good sign
            logger.info(f"  [Sci-Hub Test] Domain {domain} responsive (404 for test DOI is OK).")
            return True
        logger.warning(f"  [Sci-Hub Test] Domain {domain} failed test. Status: {response.status_code}, Content-Type: {response.headers.get('Content-Type','')}")
    except requests.exceptions.RequestException as e:
        logger.warning(f"  [Sci-Hub Test] Domain {domain} failed test: {e}")
    return False

def initialize_preferred_domains():
    global PREFERRED_DOMAINS
    logger.info("Initializing preferred Sci-Hub domains by testing...")
    # Shuffle to not always hit the same domains first
    shuffled_domains = random.sample(SCI_HUB_DOMAINS, len(SCI_HUB_DOMAINS))
    
    for domain in shuffled_domains:
        if test_scihub_domain(domain):
            PREFERRED_DOMAINS.append(domain)
            if len(PREFERRED_DOMAINS) >= 2: # Try to get at least 2 preferred domains
                break
    if not PREFERRED_DOMAINS and SCI_HUB_DOMAINS: # If no preferred found, use the first one from main list as a last resort
        if test_scihub_domain(SCI_HUB_DOMAINS[0]): # Test it first
            PREFERRED_DOMAINS.append(SCI_HUB_DOMAINS[0])
            
    if PREFERRED_DOMAINS:
        logger.info(f"Preferred Sci-Hub domains set to: {PREFERRED_DOMAINS}")
    else:
        logger.error("Could not identify any working preferred Sci-Hub domains from the list. Downloads might fail.")


def download_pdf_from_scihub_attempt(identifier, identifier_type, pmid_for_filename, first_author, year, output_dir):
    base_filename = f"{year}_{first_author}_{pmid_for_filename}"
    pdf_filename = sanitize_filename(base_filename) + ".pdf"
    pdf_filepath = os.path.join(output_dir, pdf_filename)

    if os.path.exists(pdf_filepath):
        logger.info(f"PDF already exists: {pdf_filepath} (PMID: {pmid_for_filename})")
        return True

    logger.info(f"Attempting Sci-Hub for {identifier_type}: {identifier} (PMID: {pmid_for_filename})")
    
    # Construct a list of domains to try: preferred first, then others
    domains_to_try = list(PREFERRED_DOMAINS) # Start with a copy
    remaining_domains = [d for d in SCI_HUB_DOMAINS if d not in PREFERRED_DOMAINS]
    random.shuffle(remaining_domains) # Shuffle fallbacks
    domains_to_try.extend(remaining_domains[:FALLBACK_DOMAINS_COUNT])
    
    # Ensure no duplicates if preferred are also in main list
    domains_to_try = list(dict.fromkeys(domains_to_try)) 
    if not domains_to_try: # Fallback if PREFERRED_DOMAINS was empty and SCI_HUB_DOMAINS too
        logger.error("No Sci-Hub domains available to try.")
        return False

    logger.debug(f"  Domains to try for '{identifier}': {domains_to_try}")

    for scihub_domain in domains_to_try:
        search_url = f"{scihub_domain.strip('/')}/{identifier}"
        logger.info(f"  Trying domain: {search_url}")
        try:
            time.sleep(REQUEST_DELAY_SCIHUB)
            response = session_scihub.get(search_url, timeout=30)
            response.raise_for_status()
            content_type = response.headers.get('Content-Type', '').lower()

            if 'application/pdf' in content_type:
                logger.info(f"    Direct PDF from {search_url}!")
                with open(pdf_filepath, 'wb') as f: f.write(response.content)
                logger.info(f"    SUCCESS: Downloaded {pdf_filename} (via {identifier_type} from {scihub_domain})")
                return True
            
            elif 'text/html' in content_type:
                logger.info(f"    HTML page from {search_url}. Parsing...")
                direct_pdf_url = find_pdf_url_on_scihub_page(response.content, response.url)
                if direct_pdf_url:
                    logger.info(f"    Found PDF URL: {direct_pdf_url}. Attempting download...")
                    time.sleep(0.5) # Small pause before fetching actual PDF
                    pdf_response = session_scihub.get(direct_pdf_url, timeout=60, stream=True)
                    pdf_response.raise_for_status()
                    if 'application/pdf' in pdf_response.headers.get('Content-Type','').lower():
                        with open(pdf_filepath, 'wb') as f:
                            for chunk in pdf_response.iter_content(chunk_size=8192): f.write(chunk)
                        logger.info(f"    SUCCESS: Downloaded {pdf_filename} (from {direct_pdf_url})")
                        return True
                    else:
                        logger.warning(f"    Link {direct_pdf_url} did not lead to PDF. Content-Type: {pdf_response.headers.get('Content-Type','')}")
                # else: logger_already_warned_in_find_pdf_url
            else:
                logger.warning(f"    Unexpected Content-Type '{content_type}' from {search_url}.")

        except requests.exceptions.HTTPError as e:
            if e.response.status_code == 404: logger.warning(f"    Article {identifier} not found on {scihub_domain} (404).")
            elif e.response.status_code == 403: logger.warning(f"    Access Forbidden (403) for {search_url}.")
            else: logger.error(f"    HTTP Error for {search_url}: {e.response.status_code} {e.response.reason}")
        except requests.exceptions.RequestException as e:
            logger.error(f"    Request failed for {search_url}: {e}", exc_info=False) # Less detail for common errors
        # except Exception as e: logger.error(f"    Unexpected error with {search_url}: {e}", exc_info=True)
            
    return False

def process_article_row(args):
    index, row, pmid_to_doi_map, output_dir = args
    pmid_str = str(row.get('PMID', '')).strip().split('.')[0]
    if not pmid_str:
        # logger.info(f"Skipping Excel row {index + 2}: No PMID.") # Too verbose for parallel
        return pmid_str, False
    first_author = str(row.get('First Author', 'UnknownAuthor')).strip()
    year = str(row.get('Year', 'UnknownYear')).strip()
    
    logger.info(f"Processing: PMID {pmid_str} ({first_author}, {year}) [Excel row {index+2}]")
    article_doi = pmid_to_doi_map.get(pmid_str)
    download_successful = False

    if article_doi:
        logger.info(f"  PMID {pmid_str} has DOI: {article_doi}. Trying DOI first.")
        if download_pdf_from_scihub_attempt(article_doi, "DOI", pmid_str, first_author, year, output_dir):
            download_successful = True
        else:
            logger.warning(f"  DOI ({article_doi}) download failed for PMID {pmid_str}. Fallback to PMID.")
    else:
        logger.info(f"  No DOI for PMID {pmid_str}. Trying PMID directly.")

    if not download_successful:
        if download_pdf_from_scihub_attempt(pmid_str, "PMID", pmid_str, first_author, year, output_dir):
            download_successful = True
    
    if not download_successful:
        logger.error(f"FAILED to download for PMID {pmid_str} (Excel row {index+2})")
    return pmid_str, download_successful

# --- Main Script ---
def main():
    logger.info(f"Starting PDF download process (v3)...")
    if NCBI_API_KEY == "YOUR_API_KEY_HERE": logger.warning("NCBI_API_KEY is placeholder.")
    if CROSSREF_MAILTO == "your_email@example.com": logger.warning("CROSSREF_MAILTO is placeholder.")
    
    session_ncbi.headers.update({'User-Agent': ncbi_user_agent_base.format(CROSSREF_MAILTO if CROSSREF_MAILTO != "your_email@example.com" else "anonymous_user")})

    try:
        df = pd.read_excel(EXCEL_FILE_PATH)
    except FileNotFoundError: logger.error(f"Excel file not found: {EXCEL_FILE_PATH}."); return
    except Exception as e: logger.error(f"Could not read Excel file: {e}", exc_info=True); return

    if 'PMID' not in df.columns: logger.error(f"'PMID' column not found. Cols: {df.columns.tolist()}"); return
    if not os.path.exists(OUTPUT_PDF_DIR): os.makedirs(OUTPUT_PDF_DIR); logger.info(f"Created output dir: {OUTPUT_PDF_DIR}")

    all_pmids_from_excel = [str(pmid).split('.')[0] for pmid in df['PMID'].dropna().unique() if str(pmid).strip()]
    pmid_to_doi_map = fetch_dois_for_pmids_batched(all_pmids_from_excel)
    
    initialize_preferred_domains() # Test and set preferred Sci-Hub domains

    downloaded_count = 0; failed_count = 0
    tasks = [(index, row, pmid_to_doi_map, OUTPUT_PDF_DIR) for index, row in df.iterrows()]

    logger.info(f"Starting Parallel PDF Download Phase (Workers: {MAX_WORKERS_SCIHUB}, Preferred Domains: {PREFERRED_DOMAINS})...")
    
    with ThreadPoolExecutor(max_workers=MAX_WORKERS_SCIHUB, thread_name_prefix='SciHubWorker') as executor:
        future_to_task_info = {executor.submit(process_article_row, task_args): (task_args[0], task_args[1].get('PMID')) for task_args in tasks}
        for future in as_completed(future_to_task_info):
            _excel_index, _pmid_original = future_to_task_info[future]
            try:
                _pmid_processed, success = future.result()
                if success: downloaded_count += 1
                else: failed_count += 1
            except Exception as exc:
                logger.error(f"Article (Excel idx {_excel_index}, PMID {_pmid_original}) generated exception in thread: {exc}", exc_info=True)
                failed_count += 1
                
    logger.info("--- Download Summary ---")
    logger.info(f"Total articles in Excel: {len(df)}")
    logger.info(f"Successfully downloaded/existed: {downloaded_count}")
    logger.info(f"Failed to download: {failed_count}")
    logger.info(f"PDFs saved to: {os.path.abspath(OUTPUT_PDF_DIR)}")

if __name__ == "__main__":
    # To see debug messages from find_pdf_url_on_scihub_page
    # logging.getLogger().setLevel(logging.DEBUG) 
    main()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
import time
import re
from urllib.parse import urljoin, urlparse
from urllib3.util.retry import Retry
import xml.etree.ElementTree as ET
from requests.adapters import HTTPAdapter
from concurrent.futures import ThreadPoolExecutor, as_completed
import logging
import random
from collections import Counter

# --- Configuration ---
EXCEL_FILE_PATH = r"C:\Users\Galaxy\Downloads\screening_ERAS.xlsx"
OUTPUT_PDF_DIR = "downloaded_pdfs_v4"

SCI_HUB_DOMAINS = [ # Prioritize historically better ones
    "https://sci-hub.se",
    "https://sci-hub.ru",
    "https://sci-hub.ren",
    "https://sci-hub.wf", # Problematic in logs, but kept for test diversity
    "https://sci-hub.ee",
    "https://sci-hub.st", # Problematic in logs
]
REQUEST_DELAY_SCIHUB = 1.5 # Delay between domain attempts for *same* article
MAX_WORKERS_SCIHUB = 4
PREFERRED_DOMAINS_COUNT = 1 # Try to find one solid preferred domain
FALLBACK_DOMAINS_COUNT = 2
LAST_SUCCESSFUL_PARSING_DOMAIN = None # Global to hold the last domain that allowed a successful parse

NCBI_API_KEY = "YOUR_API_KEY_HERE"
CROSSREF_MAILTO = "your_email@example.com"
EFETCH_BATCH_SIZE = 100
REQUEST_DELAY_NCBI = 0.35

# --- Logging ---
# Basic config for file and console
log_formatter = logging.Formatter('%(asctime)s - %(levelname)s - [%(threadName)s] - %(message)s')
logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO) # Overall level

# Console Handler for INFO and above
ch = logging.StreamHandler()
ch.setLevel(logging.INFO)
ch.setFormatter(log_formatter)
logger.addHandler(ch)

# File Handler for more detailed DEBUG logs (optional, can be enabled)
# fh = logging.FileHandler('pdf_downloader.log', mode='w')
# fh.setLevel(logging.DEBUG) # Capture all debugs from find_pdf_url etc.
# fh.setFormatter(log_formatter)
# logger.addHandler(fh)


# --- NCBI Session & Helper (Same as before) ---
session_ncbi = requests.Session()
retries_ncbi = Retry(total=3, backoff_factor=0.5, status_forcelist=[500, 502, 503, 504], allowed_methods=frozenset(['GET', 'POST']))
adapter_ncbi = HTTPAdapter(pool_connections=10, pool_maxsize=10, max_retries=retries_ncbi)
session_ncbi.mount("https://", adapter_ncbi)
session_ncbi.mount("http://", adapter_ncbi)
ncbi_user_agent_base = 'SciHubPDFDownloader/2.3 (Python Requests; mailto:{})'

def _get_ncbi_params(extra=None):
    params = {"tool": "pdf_downloader_eras_v4", "email": CROSSREF_MAILTO}
    if NCBI_API_KEY and NCBI_API_KEY != "YOUR_API_KEY_HERE": params["api_key"] = NCBI_API_KEY
    if CROSSREF_MAILTO == "your_email@example.com" and "email" in params:
         logger.warning("`CROSSREF_MAILTO` is placeholder.")
         params.pop("email", None)
    if extra: params.update(extra)
    return params

def fetch_dois_for_pmids_batched(pmids_list, batch_size=EFETCH_BATCH_SIZE):
    # (This function is largely the same as before, no major changes needed for efficiency here)
    t_start_fetch_doi = time.time()
    pmid_to_doi_map = {pmid: None for pmid in pmids_list}
    valid_pmids_to_query = [pmid for pmid in pmids_list if pmid]
    if not valid_pmids_to_query: return pmid_to_doi_map
    logger.info(f"Fetching DOIs for {len(valid_pmids_to_query)} PMIDs from NCBI (batch: {batch_size})")
    # ... (rest of the function remains the same, ensure it uses logger.info, logger.warning, logger.error) ...
    for i in range(0, len(valid_pmids_to_query), batch_size):
        batch_pmids = valid_pmids_to_query[i:i + batch_size]
        logger.debug(f"  NCBI EFetch batch {i//batch_size + 1} ({len(batch_pmids)} PMIDs)")
        post_data = _get_ncbi_params({"db": "pubmed", "retmode": "xml", "id": ",".join(batch_pmids)})
        try:
            response = session_ncbi.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi", data=post_data, timeout=60)
            response.raise_for_status()
            if not response.content: logger.warning(f"NCBI EFetch batch {i//batch_size+1} empty content."); continue
            root = ET.fromstring(response.content)
            for art_node in root.findall(".//PubmedArticle"):
                pmid_elem = art_node.find(".//MedlineCitation/PMID"); pmid = pmid_elem.text if pmid_elem is not None else None
                if not pmid: continue
                doi = None
                doi_elem = art_node.find(f".//ArticleIdList/ArticleId[@IdType='doi']")
                if doi_elem is None: doi_elem = art_node.find(f".//MedlineCitation/Article/ELocationID[@EIdType='doi'][@ValidYN='Y']")
                if doi_elem is not None and doi_elem.text: doi = doi_elem.text.strip()
                if pmid in pmid_to_doi_map and doi: pmid_to_doi_map[pmid] = doi
        except Exception as e: logger.error(f"NCBI EFetch batch {i//batch_size+1} error: {e}", exc_info=False)
        time.sleep(REQUEST_DELAY_NCBI if not (NCBI_API_KEY and NCBI_API_KEY != "YOUR_API_KEY_HERE") else 0.11)
    found_doi_count = sum(1 for d_val in pmid_to_doi_map.values() if d_val)
    logger.info(f"DOI fetching complete. Found for {found_doi_count}/{len(valid_pmids_to_query)} PMIDs. (Took {time.time() - t_start_fetch_doi:.2f}s)")
    return pmid_to_doi_map


# --- Sci-Hub Session & Helpers (Largely same retry, headers) ---
session_scihub = requests.Session()
scihub_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'Accept-Language': 'en-US,en;q=0.9', 'DNT': '1', 'Upgrade-Insecure-Requests': '1'
}
session_scihub.headers.update(scihub_headers)
retries_scihub = Retry(total=1, backoff_factor=0.2, status_forcelist=[500, 502, 503, 504], allowed_methods=frozenset(['GET'])) # Less retries per domain
adapter_scihub = HTTPAdapter(pool_connections=MAX_WORKERS_SCIHUB + 5, pool_maxsize=MAX_WORKERS_SCIHUB + 5, max_retries=retries_scihub)
session_scihub.mount("https://", adapter_scihub)
session_scihub.mount("http://", adapter_scihub)

def sanitize_filename(filename):
    filename = re.sub(r'[\\/*?:"<>|]', "", filename)
    filename = filename.replace(" ", "_").replace("__", "_")
    return filename[:200]

def find_pdf_url_on_scihub_page(html_content, scihub_page_url):
    # This function is CRITICAL. Its success dictates overall success.
    # Needs to be robust against Sci-Hub's changing HTML.
    soup = BeautifulSoup(html_content, 'html.parser')
    pdf_url = None
    scihub_domain_base = f"{urlparse(scihub_page_url).scheme}://{urlparse(scihub_page_url).netloc}"
    logger.debug(f"Parsing HTML from {scihub_page_url} (base: {scihub_domain_base})")

    # Primary target: iframe with id 'pdf' or src containing '.pdf' or 'downloads'
    iframe = soup.find('iframe', id='pdf')
    if iframe and iframe.get('src'):
        logger.debug(f"  Found iframe#pdf with src: {iframe['src']}")
        pdf_url = iframe['src']
    elif iframe: # Check if #pdf exists but src is missing, log it
        logger.debug("  Found iframe#pdf but it has no 'src' attribute.")


    if not pdf_url: # Fallback to iframe#article
        iframe = soup.find('iframe', id='article')
        if iframe and iframe.get('src'):
            logger.debug(f"  Found iframe#article with src: {iframe['src']}")
            pdf_url = iframe['src']

    if not pdf_url: # Fallback to any iframe if its src looks like a PDF path
        for iframe_tag in soup.find_all('iframe'):
            src_val = iframe_tag.get('src')
            if src_val and ('.pdf' in src_val or 'downloads/' in src_val or 'tree/' in src_val):
                logger.debug(f"  Found generic iframe with PDF-like src: {src_val}")
                pdf_url = src_val
                break
    
    if not pdf_url: # Try embed tag
        embed = soup.find('embed', type='application/pdf')
        if embed and embed.get('src'):
            logger.debug(f"  Found embed tag with src: {embed['src']}")
            pdf_url = embed['src']

    # Try specific buttons/links: common IDs or onclick patterns
    if not pdf_url:
        # Look for common download buttons/links by ID or attribute
        # Order of preference for selectors
        selectors = [
            'button#download', 'a#download', 'div#download', # Element with ID 'download'
            'button[onclick*="location.href"]', # Button with onclick
            'a[onclick*="location.href"]',      # Link with onclick
            'div[onclick*="location.href"]',    # Div with onclick
            'a[href*=".pdf"]',                   # Direct link to a .pdf file
            'a[href*="downloads/"]',            # Link containing 'downloads/'
            'a[href*="/tree/"]',                # Link containing '/tree/' (another Sci-Hub pattern)
        ]
        for selector in selectors:
            element = soup.select_one(selector)
            if element:
                potential_url = None
                if element.get('onclick'):
                    match = re.search(r"location\.href\s*=\s*['\"]([^'\"]+)['\"]", element['onclick'])
                    if match: potential_url = match.group(1)
                elif element.get('href') and element.get('href') != '#':
                    potential_url = element['href']
                
                if potential_url:
                    logger.debug(f"  Found element via selector '{selector}': '{element.text.strip()[:30]}' -> {potential_url}")
                    pdf_url = potential_url
                    break 
    
    if pdf_url:
        if pdf_url.startswith("//"):
            pdf_url = urlparse(scihub_page_url).scheme + ":" + pdf_url
        elif not pdf_url.startswith("http"):
            pdf_url = urljoin(scihub_domain_base, pdf_url)
        logger.debug(f"  Resolved PDF URL: {pdf_url}")
        return pdf_url
    
    logger.warning(f"  [!] Could not find PDF link on page {scihub_page_url} using defined selectors.")
    # For debugging failed parses:
    # with open(f"debug_page_{sanitize_filename(urlparse(scihub_page_url).path)}.html", "w", encoding="utf-8") as f_debug:
    #    f_debug.write(soup.prettify())
    # logger.debug(f"Saved HTML of failed parse to debug_page_{sanitize_filename(urlparse(scihub_page_url).path)}.html")
    return None


def test_scihub_domain(domain, test_identifier="10.1000/182"): # Test with a DOI
    test_url = f"{domain.strip('/')}/{test_identifier}"
    try:
        logger.debug(f"  [Domain Test] Testing {test_url}...")
        response = session_scihub.get(test_url, timeout=7) # Shorter timeout for tests
        if response.status_code == 200 and 'text/html' in response.headers.get('Content-Type','').lower():
            # Try to parse this test page - if it has a standard Sci-Hub structure, it's good.
            # Even if the specific test DOI isn't found, Sci-Hub usually shows a "not found" page
            # that itself has a structure we can check.
            if find_pdf_url_on_scihub_page(response.content, response.url) or "article not found" in response.text.lower() or "sci-hub" in response.text.lower():
                 logger.info(f"  [Domain Test] Domain {domain} PASSED (responsive & parsable/Sci-Hub structure).")
                 return True
            else:
                 logger.warning(f"  [Domain Test] Domain {domain} FAILED (responsive but HTML structure not recognized by parser).")
                 return False # Structure not parsable
        elif response.status_code == 404 and "article not found" in response.text.lower():
            logger.info(f"  [Domain Test] Domain {domain} PASSED (responsive, 404 for test DOI is acceptable Sci-Hub behavior).")
            return True
        logger.warning(f"  [Domain Test] Domain {domain} FAILED. Status: {response.status_code}, Content-Type: {response.headers.get('Content-Type','')}")
    except requests.exceptions.RequestException as e:
        logger.warning(f"  [Domain Test] Domain {domain} FAILED request: {e}")
    return False

def initialize_working_domains():
    global PREFERRED_DOMAINS_COUNT, LAST_SUCCESSFUL_PARSING_DOMAIN
    working_domains = []
    logger.info("Initializing working Sci-Hub domains by testing...")
    # Test a common, likely available DOI to check parsing capability
    # Example: DOI for "A new inhibitor of the Wnt signaling pathway, CHIR-99021, has no effect on global protein synthesis"
    test_doi_for_parsing = "10.1186/1471-2121-7-27" 

    shuffled_domains = random.sample(SCI_HUB_DOMAINS, len(SCI_HUB_DOMAINS))
    for domain in shuffled_domains:
        if test_scihub_domain(domain, test_identifier=test_doi_for_parsing): # Test with a real DOI to check parsing
            working_domains.append(domain)
            if len(working_domains) >= PREFERRED_DOMAINS_COUNT:
                break
    
    if working_domains:
        logger.info(f"Working Sci-Hub domains for initial attempts: {working_domains}")
        LAST_SUCCESSFUL_PARSING_DOMAIN = working_domains[0] # Start with the first one
    else:
        logger.error("CRITICAL: Could not identify any working Sci-Hub domains. Downloads will likely fail.")
    return working_domains


def download_pdf_from_scihub_attempt(identifier, identifier_type, pmid_for_filename, first_author, year, output_dir, available_domains):
    global LAST_SUCCESSFUL_PARSING_DOMAIN
    base_filename = f"{year}_{first_author}_{pmid_for_filename}"
    pdf_filename = sanitize_filename(base_filename) + ".pdf"
    pdf_filepath = os.path.join(output_dir, pdf_filename)

    if os.path.exists(pdf_filepath):
        # This log is now in process_article_row to reduce duplicity when called for DOI then PMID
        return True, "Already Exists"

    # Attempt with the last known good parsing domain first if it's different from current loop
    domains_to_try_ordered = []
    if LAST_SUCCESSFUL_PARSING_DOMAIN and LAST_SUCCESSFUL_PARSING_DOMAIN in available_domains:
        domains_to_try_ordered.append(LAST_SUCCESSFUL_PARSING_DOMAIN)
    
    # Add other available domains, ensuring no immediate duplication
    for d in available_domains:
        if d not in domains_to_try_ordered:
            domains_to_try_ordered.append(d)
    
    # If still empty, add some from main list as ultimate fallback
    if not domains_to_try_ordered:
        remaining_domains = [d for d in SCI_HUB_DOMAINS if d not in domains_to_try_ordered]
        random.shuffle(remaining_domains)
        domains_to_try_ordered.extend(remaining_domains[:FALLBACK_DOMAINS_COUNT + 1]) # +1 to ensure at least one if available_domains was empty
    
    domains_to_try_ordered = list(dict.fromkeys(domains_to_try_ordered)) # Ensure uniqueness
    
    logger.debug(f"  PMID {pmid_for_filename}: Trying {identifier_type} '{identifier}'. Order: {domains_to_try_ordered}")

    for scihub_domain in domains_to_try_ordered:
        search_url = f"{scihub_domain.strip('/')}/{identifier}"
        logger.debug(f"    PMID {pmid_for_filename}: Attempting {search_url}")
        try:
            time.sleep(REQUEST_DELAY_SCIHUB)
            response = session_scihub.get(search_url, timeout=20) # Shorter timeout for page load
            response.raise_for_status()
            content_type = response.headers.get('Content-Type', '').lower()

            if 'application/pdf' in content_type:
                logger.info(f"    PMID {pmid_for_filename}: Direct PDF from {search_url}!")
                with open(pdf_filepath, 'wb') as f: f.write(response.content)
                LAST_SUCCESSFUL_PARSING_DOMAIN = scihub_domain # This domain directly serves PDF
                return True, scihub_domain
            
            elif 'text/html' in content_type:
                logger.debug(f"    PMID {pmid_for_filename}: HTML from {search_url}. Parsing...")
                direct_pdf_url = find_pdf_url_on_scihub_page(response.content, response.url)
                if direct_pdf_url:
                    logger.info(f"    PMID {pmid_for_filename}: Found PDF URL: {direct_pdf_url}. Downloading...")
                    time.sleep(0.5)
                    pdf_response = session_scihub.get(direct_pdf_url, timeout=45, stream=True) # Longer for actual PDF
                    pdf_response.raise_for_status()
                    if 'application/pdf' in pdf_response.headers.get('Content-Type','').lower():
                        with open(pdf_filepath, 'wb') as f:
                            for chunk in pdf_response.iter_content(chunk_size=8192): f.write(chunk)
                        LAST_SUCCESSFUL_PARSING_DOMAIN = scihub_domain # This domain's HTML was parsable
                        logger.debug(f"      PMID {pmid_for_filename}: Parsed HTML from {scihub_domain} successfully.")
                        return True, scihub_domain
                    else:
                        logger.warning(f"    PMID {pmid_for_filename}: Link from {scihub_domain} ({direct_pdf_url}) not PDF. Type: {pdf_response.headers.get('Content-Type','')}")
                # else: find_pdf_url_on_scihub_page already logged the failure to find link
            else:
                logger.warning(f"    PMID {pmid_for_filename}: Unexpected Content-Type '{content_type}' from {search_url}.")
        except requests.exceptions.HTTPError as e:
            if e.response.status_code == 404: logger.debug(f"    PMID {pmid_for_filename}: 404 on {search_url}.")
            elif e.response.status_code == 403: logger.warning(f"    PMID {pmid_for_filename}: 403 Forbidden on {search_url}.")
            else: logger.warning(f"    PMID {pmid_for_filename}: HTTP Error {e.response.status_code} on {search_url}.")
        except requests.exceptions.RequestException as e:
            logger.warning(f"    PMID {pmid_for_filename}: Request failed for {search_url}: {e}", exc_info=False)
            
    return False, None


def process_article_row(args):
    index, row, pmid_to_doi_map, output_dir, working_domains_for_run = args
    pmid_str = str(row.get('PMID', '')).strip().split('.')[0]
    if not pmid_str: return pmid_str, False, "No PMID", None
    
    first_author = str(row.get('First Author', 'UnknownAuthor')).strip()
    year = str(row.get('Year', 'UnknownYear')).strip()
    
    article_log_prefix = f"Article (PMID {pmid_str}, {year}, {first_author}, ExcelRow {index+2})"
    logger.info(f"{article_log_prefix}: Starting process.")

    pdf_filepath = os.path.join(output_dir, sanitize_filename(f"{year}_{first_author}_{pmid_str}") + ".pdf")
    if os.path.exists(pdf_filepath):
        logger.info(f"{article_log_prefix}: PDF already exists.")
        return pmid_str, True, "Already Existed", None


    article_doi = pmid_to_doi_map.get(pmid_str)
    download_successful = False
    used_domain = None
    method = ""

    if article_doi:
        logger.info(f"{article_log_prefix}: Has DOI {article_doi}. Trying DOI first.")
        download_successful, used_domain = download_pdf_from_scihub_attempt(
            article_doi, "DOI", pmid_str, first_author, year, output_dir, working_domains_for_run)
        if download_successful: method = f"DOI via {used_domain}" if used_domain != "Already Exists" else "DOI (Exists)"
        else: logger.warning(f"{article_log_prefix}: DOI download failed. Fallback to PMID.")
    else:
        logger.info(f"{article_log_prefix}: No DOI found. Trying PMID directly.")

    if not download_successful:
        download_successful, used_domain = download_pdf_from_scihub_attempt(
            pmid_str, "PMID", pmid_str, first_author, year, output_dir, working_domains_for_run)
        if download_successful: method = f"PMID via {used_domain}" if used_domain != "Already Exists" else "PMID (Exists)"

    if download_successful:
        logger.info(f"{article_log_prefix}: SUCCESS ({method}).")
    else:
        logger.error(f"{article_log_prefix}: FAILED all attempts.")
    return pmid_str, download_successful, method, used_domain if download_successful and used_domain != "Already Exists" else None

# --- Main Script ---
def main():
    run_start_time = time.time()
    logger.info(f"Starting PDF download process (v4)...")
    if NCBI_API_KEY == "YOUR_API_KEY_HERE": logger.warning("NCBI_API_KEY is placeholder.")
    if CROSSREF_MAILTO == "your_email@example.com": logger.warning("CROSSREF_MAILTO is placeholder.")
    
    session_ncbi.headers.update({'User-Agent': ncbi_user_agent_base.format(CROSSREF_MAILTO if CROSSREF_MAILTO != "your_email@example.com" else "anonymous_user")})

    try: df = pd.read_excel(EXCEL_FILE_PATH)
    except FileNotFoundError: logger.error(f"Excel file not found: {EXCEL_FILE_PATH}."); return
    except Exception as e: logger.error(f"Could not read Excel: {e}", exc_info=True); return

    if 'PMID' not in df.columns: logger.error(f"'PMID' column not found. Cols: {df.columns.tolist()}"); return
    if not os.path.exists(OUTPUT_PDF_DIR): os.makedirs(OUTPUT_PDF_DIR); logger.info(f"Created output dir: {OUTPUT_PDF_DIR}")

    all_pmids_from_excel = [str(pmid).split('.')[0] for pmid in df['PMID'].dropna().unique() if str(pmid).strip()]
    pmid_to_doi_map = fetch_dois_for_pmids_batched(all_pmids_from_excel)
    
    # Initialize working domains ONCE for the entire run
    # These are the domains that passed initial testing and are considered most likely to work
    # for both serving pages and having a parsable structure.
    working_domains_for_this_run = initialize_working_domains()
    if not working_domains_for_this_run:
        logger.warning("No working Sci-Hub domains identified in pre-check. Using full list as fallback for each item.")
        # If no domains pass the test, each article will try a broader selection based on FALLBACK_DOMAINS_COUNT from SCI_HUB_DOMAINS
        # This is suboptimal but better than failing immediately.
        # The download_pdf_from_scihub_attempt handles this by falling back to SCI_HUB_DOMAINS if available_domains is empty.

    download_stats = Counter() # For 'success', 'failure', 'existed'
    failed_pmids = []
    domain_success_counts = Counter()

    tasks = [(index, row, pmid_to_doi_map, OUTPUT_PDF_DIR, working_domains_for_this_run) for index, row in df.iterrows()]

    logger.info(f"Starting Parallel PDF Download (Workers: {MAX_WORKERS_SCIHUB}, Initial Working Domains: {working_domains_for_this_run})...")
    
    with ThreadPoolExecutor(max_workers=MAX_WORKERS_SCIHUB, thread_name_prefix='SciHubWorker') as executor:
        future_to_info = {executor.submit(process_article_row, task_args): (task_args[0], task_args[1].get('PMID')) for task_args in tasks}
        for future in as_completed(future_to_info):
            _excel_idx, _orig_pmid = future_to_info[future]
            try:
                _pmid_proc, success, method_msg, domain_used = future.result()
                if success:
                    if method_msg == "Already Existed": download_stats['existed'] +=1
                    else: download_stats['success'] += 1
                    if domain_used: domain_success_counts[domain_used] += 1
                else:
                    download_stats['failure'] += 1
                    failed_pmids.append(_pmid_proc or _orig_pmid) # Store PMID of failed article
            except Exception as exc:
                logger.error(f"Article (Excel idx {_excel_idx}, PMID {_orig_pmid}) generated exception: {exc}", exc_info=True)
                download_stats['failure'] += 1
                failed_pmids.append(_orig_pmid)
                
    run_end_time = time.time()
    logger.info("--- Download Summary ---")
    logger.info(f"Total articles in Excel: {len(df)}")
    logger.info(f"Successfully downloaded: {download_stats['success']}")
    logger.info(f"Already existed:       {download_stats['existed']}")
    logger.info(f"Failed to download:    {download_stats['failure']}")
    if failed_pmids:
        logger.info(f"Failed PMIDs: {', '.join(filter(None,failed_pmids))}")
    if domain_success_counts:
        logger.info("Successes by domain:")
        for domain, count in domain_success_counts.items():
            logger.info(f"  {domain}: {count}")
    logger.info(f"PDFs saved to: {os.path.abspath(OUTPUT_PDF_DIR)}")
    logger.info(f"Total script execution time: {run_end_time - run_start_time:.2f} seconds.")

if __name__ == "__main__":
    # To see all DEBUG messages including HTML parsing details:
    # logging.getLogger().setLevel(logging.DEBUG) 
    # To enable file logging for debug:
    # fh = logging.FileHandler('pdf_downloader_debug.log', mode='w')
    # fh.setLevel(logging.DEBUG)
    # fh.setFormatter(log_formatter)
    # logging.getLogger().addHandler(fh)
    main()
```

```python
import pandas as pd
import requests
import os
import time
import re
import json
import xml.etree.ElementTree as ET
from urllib.parse import quote_plus, urljoin, urlparse
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
import logging
from collections import Counter
from bs4 import BeautifulSoup

# --- Configuration ---
MY_NCBI_API_KEY = "YOUR_API_KEY_HERE"
MY_EMAIL_FOR_APIS = "levi4328@gmail.com"

# --- Logging Setup ---
if not logging.getLogger(__name__).hasHandlers():
    log_formatter = logging.Formatter('%(asctime)s - %(levelname)s - [%(threadName)s] - %(funcName)s - %(message)s')
    logger = logging.getLogger(__name__)
    logger.setLevel(logging.INFO)
    logger.propagate = False
    ch = logging.StreamHandler()
    ch.setLevel(logging.INFO)
    ch.setFormatter(log_formatter)
    logger.addHandler(ch)
    # fh = logging.FileHandler('oa_downloader_v1.3_debug.log', mode='w')
    # fh.setLevel(logging.DEBUG)
    # fh.setFormatter(log_formatter)
    # logger.addHandler(fh)
    # logger.setLevel(logging.DEBUG)
else:
    logger = logging.getLogger(__name__)

# --- HTTP Sessions ---
def create_http_session(email_for_ua, tool_name="GenericTool/1.0"):
    session = requests.Session()
    retries = Retry(total=3, backoff_factor=0.7, status_forcelist=[429, 500, 502, 503, 504],
                    allowed_methods=frozenset(['GET', 'POST'])) # Increased backoff slightly
    adapter = HTTPAdapter(max_retries=retries, pool_connections=15, pool_maxsize=30) # Increased pool
    session.mount("http://", adapter)
    session.mount("https://", adapter)
    session.headers.update({
        'User-Agent': f'{tool_name} (mailto:{email_for_ua}; for academic research)',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,application/pdf,*/*;q=0.8', # Added application/pdf
        'Accept-Language': 'en-US,en;q=0.5'
    })
    return session

session_http = create_http_session(MY_EMAIL_FOR_APIS, "OAFinderHTTP/1.3")
session_ncbi_alt = create_http_session(MY_EMAIL_FOR_APIS, "OAFinderNCBI/1.3")

def _get_ncbi_params_alt(api_key_for_ncbi, extra=None):
    params = {"tool": "oa_pdf_finder_v4", "email": MY_EMAIL_FOR_APIS} # Hardcoded email now
    if api_key_for_ncbi and api_key_for_ncbi != "YOUR_API_KEY_HERE":
        params["api_key"] = api_key_for_ncbi
    if extra: params.update(extra)
    return params

def sanitize_filename_alt(filename_str):
    if not isinstance(filename_str, str): filename_str = str(filename_str)
    s = re.sub(r'[\\/*?:"<>|]', "", filename_str)
    s = s.replace("\n", " ").replace("\r", "").replace("/", "_").replace(":", "_")
    s = re.sub(r'\s+', '_', s).strip('_')
    return s[:180] # Slightly longer filenames allowed


def download_pdf_from_url(pdf_url, filepath, pmid_for_log="N/A", source_description="Source", current_session=None):
    downloader_session = current_session if current_session else session_http
    logger.info(f"PMID {pmid_for_log}: Attempting download ({source_description}): {pdf_url}")
    try:
        response = downloader_session.get(pdf_url, timeout=(10, 60), stream=True, allow_redirects=True)
        response.raise_for_status()
        # Log final URL and content type *before* checking it
        final_url_for_download = response.url
        final_content_type = response.headers.get('Content-Type', '').lower()
        logger.debug(f"PMID {pmid_for_log}: Final download URL: {final_url_for_download}, Content-Type: {final_content_type}")

        # Even if initial headers were misleading, if the final URL ends with .pdf, and we got a 200,
        # it's highly likely a PDF. The actual check for 'application/pdf' is good,
        # but we also need to trust the URL structure for direct PDF links from PMC.
        if 'application/pdf' not in final_content_type and final_url_for_download.lower().endswith('.pdf'):
            logger.warning(f"PMID {pmid_for_log}: Content-Type is '{final_content_type}' but URL '{final_url_for_download}' ends with .pdf. Proceeding with download attempt.")
            # Proceed to download, the binary content will be the ultimate decider.
        elif 'application/pdf' not in final_content_type:
            logger.error(f"PMID {pmid_for_log}: URL {final_url_for_download} ({source_description}) not direct PDF. Content-Type: {final_content_type}.")
            return False, "Not a direct PDF stream"

        with open(filepath, 'wb') as f:
            for chunk in response.iter_content(chunk_size=65536): # Larger chunk
                f.write(chunk)
        
        file_size = os.path.getsize(filepath) if os.path.exists(filepath) else 0
        if file_size > 1000: # Greater than 1KB
            logger.info(f"PMID {pmid_for_log}: SUCCESS! Downloaded to {filepath} from {source_description} (Size: {file_size} B).")
            return True, f"Downloaded from {source_description}"
        else:
            logger.error(f"PMID {pmid_for_log}: Downloaded file {filepath} from {source_description} is empty or too small (Size: {file_size} B). Deleting.")
            if os.path.exists(filepath): os.remove(filepath)
            return False, "Empty or too small file"

    except requests.exceptions.HTTPError as e:
        logger.error(f"PMID {pmid_for_log}: HTTP Error {e.response.status_code} from {pdf_url} ({source_description})")
        return False, f"HTTP Error {e.response.status_code}"
    # Add other specific exceptions if needed (Timeout, ConnectionError etc.)
    except requests.exceptions.RequestException as e:
        logger.error(f"PMID {pmid_for_log}: Request Exception for {pdf_url} ({source_description}): {e}")
        return False, f"Request Exception"
    except Exception as e:
        logger.error(f"PMID {pmid_for_log}: Unexpected error downloading {pdf_url} ({source_description}): {e}", exc_info=True)
        return False, "Unexpected download error"


def get_oa_pdf_from_unpaywall(doi):
    if not doi: logger.warning("Unpaywall: No DOI provided."); return None
    api_url = f"https://api.unpaywall.org/v2/{quote_plus(doi)}?email={MY_EMAIL_FOR_APIS}"
    logger.debug(f"Unpaywall: Querying {api_url}")
    try:
        response = session_http.get(api_url, timeout=20)
        if response.status_code == 404: logger.info(f"Unpaywall: DOI {doi} not found."); return None
        if response.status_code == 422: logger.error(f"Unpaywall: HTTP 422 Unprocessable Entity for DOI {doi}."); return None
        response.raise_for_status() # For other HTTP errors
        data = response.json()
        if data and data.get("is_oa"):
            pdf_url_to_try = None
            best_oa_location = data.get("best_oa_location")
            if best_oa_location and best_oa_location.get("url_for_pdf"):
                pdf_url_to_try = best_oa_location.get("url_for_pdf")
                logger.info(f"Unpaywall: Found PDF URL (best_oa) for DOI {doi}: {pdf_url_to_try}")
            else: # Check all oa_locations
                for loc in data.get("oa_locations", []):
                    if loc.get("url_for_pdf"):
                        pdf_url_to_try = loc.get("url_for_pdf")
                        logger.info(f"Unpaywall: Found PDF URL (oa_locations) for DOI {doi}: {pdf_url_to_try}")
                        break # Take the first one from oa_locations
            if pdf_url_to_try: return pdf_url_to_try
            logger.info(f"Unpaywall: DOI {doi} is OA, but no PDF URL in best_oa or oa_locations.")
        else: logger.info(f"Unpaywall: No OA or no PDF URL for DOI {doi}. Free text: {data.get('free_fulltext_url', 'N/A')}")
    except Exception as e: logger.error(f"Unpaywall: Error for DOI {doi}: {e}", exc_info=False)
    return None


def get_pmcid_from_pmid(pmid, ncbi_api_key):
    # (This function is generally okay, ensure it logs appropriately)
    if not pmid: return None
    logger.debug(f"PMC: Getting PMCID for PMID {pmid}.")
    params = _get_ncbi_params_alt(ncbi_api_key, {"dbfrom": "pubmed", "db": "pmc", "id": pmid, "cmd": "neighbor_score"})
    linknames_to_try = ["pubmed_pmc_refs", "pubmed_pmc"]
    for linkname in linknames_to_try:
        params["linkname"] = linkname
        try:
            response = session_ncbi_alt.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/elink.fcgi", data=params, timeout=15)
            response.raise_for_status()
            root = ET.fromstring(response.content)
            link_ids = root.findall("./LinkSet/LinkSetDb/Link/Id")
            if link_ids:
                pmcid_text = link_ids[0].text
                if pmcid_text:
                    pmcid = pmcid_text if pmcid_text.upper().startswith("PMC") else "PMC" + pmcid_text.strip()
                    logger.info(f"PMC: Found PMCID {pmcid} for PMID {pmid} (linkname: {linkname}).")
                    return pmcid
        except Exception as e:
            logger.warning(f"PMC: Error getting PMCID for PMID {pmid} (linkname {linkname}): {e}", exc_info=False)
            if isinstance(e, requests.exceptions.HTTPError) and response and "Cannot ELink from db" in response.text: break
    logger.info(f"PMC: No PMCID found for PMID {pmid} via elink.")
    return None


def download_from_pmc(pmcid, filepath, pmid_for_log):
    if not pmcid: return False, "No PMCID provided"
    base_article_page_url = f"https://www.ncbi.nlm.nih.gov/pmc/articles/{pmcid}/pdf/"
    logger.info(f"PMC: Attempting access for {pmcid} via initial URL: {base_article_page_url}")

    try:
        pmc_session = create_http_session(MY_EMAIL_FOR_APIS, "OAFinderPMC/1.3_PMC")
        response = pmc_session.get(base_article_page_url, timeout=(10, 30), allow_redirects=True)
        response.raise_for_status()
        
        final_url_after_redirects = response.url
        final_content_type = response.headers.get('Content-Type', '').lower()
        logger.info(f"PMC: For {pmcid}, initial URL led to final URL: {final_url_after_redirects} (Content-Type: {final_content_type})")

        # CRITICAL CHANGE: If the final URL *is* a PDF link (ends with .pdf), try to download it directly.
        if final_url_after_redirects.lower().endswith('.pdf'):
            logger.info(f"PMC: Final URL for {pmcid} appears to be a direct PDF link. Attempting download.")
            return download_pdf_from_url(final_url_after_redirects, filepath, pmid_for_log, f"PubMed Central (Direct from {pmcid} redirect)", current_session=pmc_session)
        
        # If not a direct PDF link after redirects, then check if it's HTML to parse
        elif 'text/html' in final_content_type:
            logger.info(f"PMC: Final URL {final_url_after_redirects} is HTML for {pmcid}. Parsing for specific PDF link.")
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # More targeted selectors for links within the PMC HTML page
            link_selectors = [
                'ul.format-menu li a[href$=".pdf"]',            # Standard format menu
                'div.fm-sec-pdf a[href$=".pdf"]',                # PDF section in format menu
                'div.format-dropdown ul li a[href$=".pdf"]',     # Dropdown format menu
                'a.pdf-link[href$=".pdf"]',                      # Explicit class
                'a[title*="Download PDF"][href$=".pdf"]',        # Title attribute
                f'a[href*="/pdf/"][href*="{pmcid.replace("PMC","")}"][href$=".pdf"]', # Contains PMCID num & /pdf/
                'a[href*="/articles/{pmcid}/pdf/"]' # A link that itself contains the full base path again + filename
            ]
            actual_pdf_link_on_page = None
            for selector in link_selectors:
                link_tag = soup.select_one(selector)
                if link_tag and link_tag.get('href'):
                    href_val = link_tag.get('href')
                    # Ensure it's a valid-looking link
                    if href_val and not href_val.startswith("javascript:"):
                        resolved_url = urljoin(final_url_after_redirects, href_val)
                        if resolved_url.lower().endswith(".pdf"): # Double check
                             logger.info(f"PMC: Found PDF link on landing page for {pmcid} via selector '{selector}': {resolved_url}")
                             actual_pdf_link_on_page = resolved_url
                             break
            
            if actual_pdf_link_on_page:
                return download_pdf_from_url(actual_pdf_link_on_page, filepath, pmid_for_log, f"PubMed Central (Parsed from {pmcid} landing)", current_session=pmc_session)
            else:
                logger.warning(f"PMC: No definitive PDF link found on HTML landing page for {pmcid} ({final_url_after_redirects}).")
                # Optional: save HTML
                # debug_html_path = os.path.join(os.getcwd(), f"debug_pmc_{pmcid}_landing.html")
                # with open(debug_html_path, "w", encoding="utf-8") as f_html: f_html.write(soup.prettify())
                # logger.debug(f"Saved PMC HTML for {pmcid} to {debug_html_path}")
                return False, "PMC: PDF link not found in HTML"
        else:
            logger.warning(f"PMC: Unexpected Content-Type '{final_content_type}' for {pmcid} at {final_url_after_redirects} (not PDF or HTML).")
            return False, f"PMC: Unexpected Content-Type {final_content_type}"

    except requests.exceptions.HTTPError as e:
        logger.error(f"PMC: HTTP Error {e.response.status_code} for {pmcid} (Initial URL: {base_article_page_url})")
        return False, f"PMC HTTP Error {e.response.status_code}"
    except Exception as e:
        logger.error(f"PMC: Error processing {pmcid}: {e}", exc_info=True)
        return False, "PMC processing error"


# --- Main Orchestration ---
def attempt_oa_downloads(articles_info, download_dir, ncbi_api_key):
    # (Main logic structure of this function remains similar, calls the updated download_from_pmc)
    if not os.path.exists(download_dir): os.makedirs(download_dir)
    successful_downloads, already_existed_count = 0, 0
    failed_oalookup_items = [] # Store dicts of failed items for more info
    download_sources = Counter()

    for i, article in enumerate(articles_info):
        pmid = article.get("pmid")
        doi = article.get("doi")
        title_for_file = article.get("title_for_file", doi or pmid or f"unknown_article_{i+1}")
        filename_stem = sanitize_filename_alt(title_for_file)
        pdf_filepath = os.path.join(download_dir, filename_stem + ".pdf")

        logger.info(f"\n--- OA Check {i+1}/{len(articles_info)} for PMID: {pmid}, DOI: {doi or 'N/A'} ---")
        if os.path.exists(pdf_filepath):
            logger.info(f"PDF already exists: {pdf_filepath}. Skipping."); already_existed_count += 1; continue

        downloaded_this_article = False; source_of_download = "Unknown"; status_msg = "Not found"

        if doi: # Try Unpaywall first if DOI exists
            unpaywall_pdf_url = get_oa_pdf_from_unpaywall(doi)
            if unpaywall_pdf_url:
                success, status_msg = download_pdf_from_url(unpaywall_pdf_url, pdf_filepath, pmid, "Unpaywall")
                if success: downloaded_this_article = True; source_of_download = "Unpaywall"
        
        if not downloaded_this_article and pmid: # Then try PMC if PMID exists
            pmcid = get_pmcid_from_pmid(pmid, ncbi_api_key)
            if pmcid:
                success, status_msg = download_from_pmc(pmcid, pdf_filepath, pmid)
                if success: downloaded_this_article = True; source_of_download = "PubMed Central"
        
        if downloaded_this_article:
            successful_downloads += 1
            download_sources[source_of_download] += 1
        else:
            logger.info(f"PMID {pmid}: No OA PDF found/downloaded. Last status: {status_msg}")
            failed_oalookup_items.append({'pmid': pmid, 'doi': doi, 'status': status_msg})

    logger.info("\n--- Open Access Download Summary ---")
    logger.info(f"Total articles for OA check: {len(articles_info)}")
    logger.info(f"Successfully downloaded via OA: {successful_downloads}")
    logger.info(f"Already existed (skipped): {already_existed_count}")
    if download_sources:
        logger.info("Downloads by OA source:"); [logger.info(f"  {s}: {c}") for s,c in download_sources.items()]
    if failed_oalookup_items:
        logger.info(f"Items for which no OA PDF was found/downloaded ({len(failed_oalookup_items)}):")
        for item in failed_oalookup_items: logger.info(f"  PMID: {item['pmid']}, DOI: {item['doi']}, Status: {item['status']}")
    return successful_downloads, already_existed_count, failed_oalookup_items


if __name__ == "__main__":
    logger.info(f"Using email: {MY_EMAIL_FOR_APIS} for API calls.")
    if MY_NCBI_API_KEY == "YOUR_API_KEY_HERE": logger.warning("NCBI_API_KEY is placeholder.")

    example_failed_pmids = [
        "39955421", "40340819", "39520824", "39489669", "39384309", "39185540", 
        "39068053", "38673038", "37802689", "39083294", "36788057", "39342249", 
        "37062759", "36969299", "35790215", "31274269", "33401363", "26888001"
    ]
    example_pmid_to_doi_map = {
        "39955421": "10.1007/s00383-025-05977-0", "40340819": "10.1136/bmjpo-2024-003280",
        "39520824": "10.1016/j.jpedsurg.2024.162046", "39489669": "10.1053/j.jvca.2024.10.005",
        "39384309": "10.1136/bmjpo-2024-002824", "39185540": "10.14309/crj.0000000000001469",
        "39068053": "10.1016/j.jpedsurg.2024.06.021", 
        "38673038": "10.3390/jpm14040411", # This one (PMC11051180) should work now
        "36788057": None, # This one (PMC11680611) should work now
        "37802689": "10.1053/j.jvca.2023.09.006", # Test DOI if available, example here
        # Fill other DOIs if available, otherwise None
        "39083294": None, "39342249": "10.1186/s12887-024-05083-5", # Example
        "37062759": "10.1089/lap.2022.0537", # Example
        "36969299": "10.3389/fped.2023.933158", # Example
        "35790215": "10.4097/kja.22279", # Example
        "31274269": "10.23736/S0375-9393.19.13880-1", # Example
        "33401363": "10.2196/10996", # Example
        "26888001": None
    }
    example_pmid_to_title_map = {
        "39955421": "Li_S_2025_Pneumonectomies", "40340819": "Pentz_B_2025_ERAS_Pain",
        "39520824": "Mansfield_SA_2025_ERAS_Neonates", "39489669": "Meier_S_2025_JVCA_ERAS",
        "39384309": "Pilkington_M_2024_BMJPO_Pain", "39185540": "Pilkington_M_2024_CRJ_Esophageal",
        "39068053": "Pilkington_M_2024_JPedSurg_Gastroschisis", "38673038": "Zacha_S_2024_JPM_Anesthesia",
        "37802689": "Alfaras-Melainis_K_2024_JVCA_Postoperative", "39083294": "Wishahi_M_2024_JAMA_Surg_AI",
        "36788057": "Downing_L_2023_JPedSurg_Telemedicine", "39342249": "Guo_R_2024_BMC_Peds_Appendicitis",
        "37062759": "Martynov_I_2023_LAP_Hernia", "36969299": "Huang_J_2023_Front_Peds_Pyloromyotomy",
        "35790215": "Lucente_M_2022_KJA_Analgesia", "31274269": "Paladini_G_2019_Minerva_Anestesiol_ERAS",
        "33401363": "Wildemeersch_D_2018_JMIR_Mobile_Health", "26888001": "Wang_Y_2015_ERAS_Thoracic"
    }

    articles_to_check_oa = []
    for pmid in example_failed_pmids:
        doi = example_pmid_to_doi_map.get(pmid)
        title_stub = example_pmid_to_title_map.get(pmid, pmid) 
        articles_to_check_oa.append({"pmid": pmid, "doi": doi, "title_for_file": title_stub})

    OA_DOWNLOAD_DIR = os.path.join(os.getcwd(), "oa_downloads_v4_final_pmc_test") # New dir for test
    attempt_oa_downloads(
        articles_info=articles_to_check_oa,
        download_dir=OA_DOWNLOAD_DIR,
        ncbi_api_key=MY_NCBI_API_KEY
    )
    logger.info("OA download process finished.")
```
